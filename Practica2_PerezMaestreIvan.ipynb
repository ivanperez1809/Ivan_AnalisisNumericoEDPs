## **Apartado 2. Introducción**

**A lo largo de esta práctica vamos a resolver la ecuación del calor para un medio isótropo y homogéneo en un dominio acotado $\Omega \subset \mathbb R^n$:**

**<center>$u_t(x,t) - \Delta u(x,t) = f(x,t)$ $\forall (x,t) \in \Omega \times (0,T)$ (1)</center>**
    
**donde $f \in C^0\left(\bar{\Omega}\times [0,T]\right)$, como ya vimos en las clases de teoría**.

**Cuando el medio no es isótropo, la ecuación del calor toma otra forma, por ejemplo, en dos dimensiones espaciales podría tomar la forma:**

**<center>$u_t(x,y,t) - k_1u_{xx}(x,y,t) - k_2u_{yy}(x,y,t) = f(x,y,t)$ $\forall (x,y,t) \in \Omega \times (0,T)$ (2)</center>**

**donde $k_1,k_2 > 0$ son constantes características del medio.**

#### **a) Busca un cambio de coordenadas lineal de manera que la ecuación (2) se transforme en (1). ¿Cómo se transforma el dominio?**

Sin más que tener en cuenta cómo son ambas ecuaciones (aparecen segundas derivadas) y tener en cuenta la linealidad de estas derivadas, si se toma el siguiente cambio de coordenadas (dejando la coordenada temporal tal y como está):

<center>$(x,y)\longmapsto (x',y') = \left(\frac{x}{\sqrt{k_1}},\frac{y}{\sqrt{k_2}}\right)$</center>
<br>

Se alcanza lo que se busca ya que, por dejarlo claro, por ejemplo: $u_{xx}(x,y,t) = \frac{1}{k_1}u_{x'x'}(x',y',t)$ sin más que aplicar la regla de la cadena. Además, comentar que dicha transformación tiene sentido pues $k_1, k_2 > 0$ por hipótesis. De esta forma, efectivamente se consigue que se tenga la siguiente ecuación en las nuevas coordenadas:

<center>$u_t(x',y',t) - \Delta u(x',y',t) = f(x',y',t)$</center>
<br>

Para analizar cómo le afecta esto al dominio, simplemente tengamos en cuenta que lo que se está haciendo es estirar o contraer la parte espacial de las coordenadas ($\Omega$) por sendos factores $1/\sqrt{k_1}, 1/\sqrt{k_2}$. De esta forma, si por ejemplo $k_1,k_2 < 1$ se tiene un "aumento" del espacio, es decir, un estiramiento. También podría pasar que una coordenada se estire y otra se contraiga, que las dos se contraigan, todo es análogo y depende de los valores de $k_1,k_2$.

#### **b) Supongamos por simplicidad una única variable espacial. Justifica que la solución para $k > 0$ y $u_0\in C_C^{\infty}(\mathbb R)$ de:**

#### **<center>$u_t(x,t) - ku_{xx}(x,t) = 0$ : $x\in\mathbb R, t > 0$</center>**
#### **<center>$u(x,0) = u_0(x)$ : $x\in\mathbb R$</center>**

#### **viene dada por:**

#### **<center>$u(x,t) = \frac{1}{\sqrt{4\pi tk}}\int_{\mathbb R}e^{-\frac{(x-y)^2}{4tk}}u_0(y)dy$</center>**

#### **Concluye que la difusión es más rápida si $k$ es mayor.**

Una de las posibles ideas, como bien se indica en la pista del enunciado, es utilizar el cambio de variable que se ha visto del caso anterior en 2D y aplicarlo en este caso unidimensional, es decir, tomar $x \longmapsto x' = \frac{x}{\sqrt{k}}$ y dejando la coordenada temporal como estaba. De esta forma, nuestra ecuación a resolver pasa a ser, como ya se adelantó en el apartado a):

<center>$u_t(x',t) - u_{x'x'}(x',t) = 0$</center>
<br>

Por supuesto, también se ve afectado el dato inicial por dicho cambio de variable, quedando: $u(x',0) = u_0(x') = u_0(\frac{x}{\sqrt{k}})$.

Démonos cuenta que, en este caso, el dominio de nuestra ecuación sigue siendo $\mathbb R$ puesto que, como se dijo antes, dicho cambio de variable supone contraer o estirar nuestro espacio, pero como el dominio inicial era $\mathbb R$, esto no supone ningún problema. Por ello, y dado que en clases de teoría ya hemos visto cuál es la solución a la ecuación del calor en $\mathbb R$ (núcleo del calor), podemos asegurar que la solución a la ecuación anterior viene dada por:

<center>$u(x',t) = \frac{1}{\sqrt{4\pi t}}\int_{\mathbb R} e^{-\frac{(x'-y')^2}{4t}}u_0(y')dy'$</center>
<br>

Pongo $y'$ como variable muda de integración para hacer claro que se va a tener que hacer un cambio de variable muy sencillo (recuperar el que se hizo inicialmente) para obtener la solución que nosotros buscamos. Dicho cambio de variable en la integral, como ya se ha dicho, es el evidente y es recuperar el espacio "tal y como lo teníamos", es decir, usar que $x' = \frac{x}{\sqrt{k}}$ y hacer el cambio de variable $y' = \frac{y}{\sqrt{k}}$ y, por tanto, $dy' = \frac{1}{\sqrt{k}}dy$.

Sin más que realizar dicho cambio de variable y agrupar en la exponencial, se obtiene que, efectivamente, la solución de nuestro problema original es:

<center>$u(x,t) = \frac{1}{\sqrt{4\pi tk}}\int_{\mathbb R}e^{-\frac{(x-y)^2}{4tk}}u_0(y)dy$</center>
<br>

Veamos ahora lo último que se pide, es decir, que la difusión del calor es más rápida a medida que se aumenta el valor de $k$:

Esencialmente, para esto, simplemente comento que, obviamente tiene que cambiar la difusión del calor por el factor $k$ que acompaña ahora en todo momento al tiempo tanto fuera de la integral como en la exponencial de dentro.

## **Apartado 3. Principios de comparación**

**Nosotros estudiamos en clase de teoría el principio del máximo cuando se tenían condiciones de Dirichlet, así como los principios de comparación que este implicaba. Estos principios se pueden generalizar (ver _Theorem 3.0.1_)**

#### **c) Prueba que para condiciones frontera homogéneas (es decir, $g^D = g^N = g^R = 0$) y mismo término de creación $f \geq 0$, las soluciones de los problemas de Dirichlet, Robin y Neumann para una misma condición inicial $u_0 \geq 0$ están ordenadas y se tiene que $0\leq u_D\leq u_R \leq u_N$. Da una interpretación física a que $u_D \leq u_N$ de acuerdo a la interpretación que se dio en clase de dichas condiciones frontera.**

**1) Veamos que $u^D, u^N, u^R \geq 0$:**

En primer lugar, es muy útil para lo que sigue hacer una observación bastante trivial y que se obtiene como consecuencia directa del teorema del enunciado. Tomemos el siguiente problema:

<center>$u_t(x,t) - \Delta u(x,t) = 0$ : $x \in \Omega$, $t \in (0,T)$</center>
<center>$u(x,0) = 0$ : $x \in \Omega$</center>
<center>$Bu (x,t) = 0$ : $x\in \partial\Omega$</center>
<br>

Es decir, consideramos el problema que estamos tratando en todo problema pero con término de creación, condición inicial y condiciones de contorno nulos. Démonos cuenta que una solución de dicho problema es la solución constantemente nula ($u(x,t) \equiv 0$). Para esta observación solo queda darse cuenta de que, en nuestro problema original, suponemos que, por hipótesis $f, g, u_0 \geq 0$, lo cual, sin más que aplicar _theorem 3.0.1_ hace que las soluciones a los tres problemas (Neumann, Dirichlet, Robin) sean mayores o iguales que 0, es decir, ya se tiene garantizada una de las cosas que se piden:

<center>$u^N, u^D, u^R \geq 0$ $\forall (x,t) \in \bar{\Omega} \times [0,T]$</center>
<br>

Ahora hay que centrarse en ver que, efectivamente, las soluciones están ordenadas.

**2) Veamos que $u^R \geq u^D$:**

Para ello razonamos como suele ser común en casos en los que se quieren probar este tipo de casos. Tomemos $v = u^R - u^D$ y se comprueba que se satisface que:

<center>$v_t - \Delta v = 0$ : $x\in \Omega$, $t \in (0,T)$</center>
<center>$v(x,0) = 0$ : $x \in \Omega$</center>
<center>$v(x,t) \geq 0$ : $x \in \partial\Omega$</center>
<br>

Las dos primeras se deducen de manera directa de aplicar la ecuación a $v$ y de que en ambos casos (Robin y Dirichlet) los términos de creación y condición inicial son exactamente los mismos (luego restando, todo se anula).

Para las condiciones de frontera hay que pararse quizá un poco más y simplemente observar que $v(x,t) = u^R(x,t) - u^D(x,t), x\in \partial\Omega$ y que, las condiciones de contorno Dirichlet hacen que $u^D$ se anule en $\partial\Omega$ mientras que, como se ha visto antes, se tiene garantizado que $u^R \geq 0 \forall x \in \partial\Omega$, luego se tiene la desigualdad que se mostraba.

En esta situación se puede contrastar el problema que se tiene para $v(x,t)$ y el que se tenía con la solución trivial y, de nuevo, sin más que aplicar _theorem 3.0.1_ se tiene que:
<center>$v(x,t) \geq 0 \Longrightarrow u^R(x,t) \geq u^D(x,t)$</center>
<br>

**3) Veamos que $u^N \geq u^R$:**

El razonamiento es exactamente el mismo que en el caso anterior. Definiendo $w(x,t) = u^N(x,t) - u^R(x,t)$ se tiene que el problema asociado a $w$ es:

<center>$w_t - \Delta w = 0$ : $x\in \Omega$, $t\in (0,T)$</center>
<center>$w(x,0) = 0$ : $x\in\Omega$</center>
<center>$\frac{\partial w}{\partial n} \geq 0$ : $x \in \partial\Omega$</center>
<br>

Antes de justificar un poco esto, se hace la deducción que se quería y es que, comparando con la solución trivial de nuevo, se deduce que:

<center>$w(x,t) \geq 0 \Longrightarrow u^N \geq u^R$</center>
<br>

Para dar consistencia a esto, se hace algún comentario. De nuevo, la ecuación y condición inicial son triviales, pero hay que pararse a comentar las condiciones de contorno en este caso.

Observemos que, en $\partial\Omega$, por cómo son $u^N$ y $u^R$ se tiene $\frac{\partial w}{\partial n} = \frac{\partial u^N}{\partial n} - \frac{\partial u^R}{\partial n} = -\frac{\partial u^R}{\partial n}$ pues todas las condiciones de frontera son nulas, luego la parte de Neumann es 0. Ahora usemos que también $u^R = 0, x\in \partial\Omega$ (por hipótesis $\alpha, \beta \geq 0$):

<center>$\alpha u^R + \beta\frac{\partial u^R}{\partial n} = 0$ $\Longrightarrow$ $\frac{\partial u^R}{\partial n} = -\frac{\alpha}{\beta}u^R$ $\Longrightarrow$ $\frac{\partial u^R}{\partial n} \leq 0$</center>
<br>

Pues, por lo probado al principio, $u^R \geq 0, x\in \partial\Omega$. Con esto, ya se comprueba que, efectivamente se cumple la condición de contorno que se había mostrado.

<br>
En cuanto a la explicación física que se le puede dar al hecho de que $u_D \leq u_N$ es bastante sencilla: ya se discutió en clase que la condición Dirichlet es básicamente como poner hielo a 0ºC en los extremos y la Neumann es, esencialmente, aislar la barra, con lo que en la primera situación se escapa calor y en la segunda no, eso explicaría que con Neumann el calor mayore a la situación con Dirichlet.

## **Apartado 4. La masa total.**

**Nosotros vamos a solucionar numéricamente el problema con condiciones homogéneas Neumann en un intervalo acotado, es decir:**

**<center>$u_t(x,t) - u_{xx}(x,t) = f(x,t)$ $\forall (x,t) \in (0,1)\times (0,T)$</center>**
**<center>$u(x,0) = u_0(x)$ $\forall x\in (0,1)$</center>**
**<center>$\frac{\partial u}{\partial x}(0,t) = \frac{\partial u}{\partial x}(1,t) = 0$ $\forall t \in [0,T]$</center>**
<br>

**Resolveremos el problema considerando como término fuente la función:**

**<center>$f(x) = S(x,t) = (1+cos(t))x(1-x)$ si $t\leq \pi$</center>**
**<center>$f(x) = 0$ si $t > \pi$</center>**
<br>

**y como la condición inicial: $u_0(x) = x^2(1-x)^2$**




#### **e) Calcula analíticamente la masa total en función del tiempo $M(t) = \int_0^1u(x,t)dx$. Haz una gráfica de $M(t)$.**

Usando la indicación que se aporta en el enunciado, integremos la ecuación: $\int_0^1u_t(x,t)dx - \int_0^1 u_{xx}(x,t)dx = \int_0^1 f(x,t)dx$.

Ahora tengamos en cuenta lo siguiente, en primer lugar, usamos el teorema fundamental del cálculo para la segunda integral de $u_{xx}$ u, en segundo lugar, sacamos la derivada temporal de la primera integral (quizá este último paso requiera de pararse a pensarlo un poco más ya que no es trivial que eso se pueda hacer, pero no ahondo en ello). De esta forma se llega a que:

<center>$\frac{d}{dt}\int_0^1u(x,t)dx - u_x(x,t)|_0^1 = \int_0^1 f(x,t)dx$ </center>
<br>

Ahora bien, teniendo en cuenta (de nuevo con la pista del enunciado) cuáles son nuestras condiciones de frontera y la definición de la masa del sistema, se llega finalmente a que:

<center>$M'(t) = \int_0^1 f(x,t)dx$</center>
<br>

Sin más que volver a integrar a ambos lados, se obtiene una expresión para la masa del sistema en función del tiempo:

<center>$M(t) - M(0) = \int_0^t\int_0^1f(x,s)dxds$</center>
<br>

Con esto, hay que calcular dicha integral (donde obsérvese que a partir de $t = \pi$ la integral ya no cambia puesto que $f$ se anula y, por supuesto, $M(0) = \int_0^1 u_0(x)$ 

```python
#Calculo ambas integrales que se necesitan con sympy (librería de cálculo simbólico de Python):
x = sp.symbols('x') #Se toma x como una variable simbólica y análogo para las siguientes
s = sp.symbols('s')
t = sp.symbols('t')

#Hago las integrales con la función integrate de sympy
M_0 = sp.integrate(x**2*(1-x)**2, (x,0, 1))
integral_doble = sp.integrate((1+sp.cos(s))*x*(1-x),(x,0,1),(s,0,t))
integral_doble + M_0 #Con esto se muestra como salida justo la solución para M(t)
```

Con este resultado tan sencillo se deduce que:

<center>$M(t) = \frac{1}{6}\left(t + sen(t)\right) + \frac{1}{30}$ si  $t\leq \pi$</center>
<center>$M(t) = \frac{\pi}{6} + \frac{1}{30}$ si  $t > \pi$</center>
<br>

A continuación, hacemos un plot de la masa del sistema como se pide.

```python
x1 = np.linspace(0,np.pi,100)
x2 = np.linspace(np.pi, 2*np.pi, 100) #Como no tenemos un valor concreto de T, ploteo hasta el doble del "cambio" en la f por ejemplo. Podría...
                                      #...tomar T = 5 que es el que se usa después, pero da exactamente igual.

M1_anali = x1/6 + np.sin(x1)/6 + 1/30
M2_anali = (np.pi/6 + 1/30) * np.ones(len(x2)) #No es más que una función continua

plt.plot(x1,M1_anali, color = 'blue', label = 't$\leq\pi$')
plt.plot(x2,M2_anali, color = 'red', label = 't$>\pi$')

plt.xlabel('t')
plt.ylabel('M(t)')
plt.title('Masa del sistema')
plt.legend()
plt.show()
```

Por comentar un poco lo que se observa, se ve que una vez el término de producción $f$ se anula, la masa del sistema se vuelve constante. Esencialmente esto se debe a que como se tienen condiciones Neumann en la frontera (que vimos que sería como poner aislantes térmicos en los extremos de nuestra barra), si el término de producción es 0, la masa no tiene ninguna forma de cambiar, es decir, se mantiene constante.

## **Apartado 5. Planteamiento numérico.**

**Vamos a plantear ahora el problema en el marco numérico. Para ello, consideraremos una discretización de $N+1$ puntos del intervalo $[0,1]$. Así, $\Delta x = 1/N$. Además, discretizaremos el intervalo temporal $[0,T]$ en $M+1$ puntos, así, $\Delta t = T/M$**

#### **f) Escribe el esquema $\theta$ asociado al problema anterior. Discute los pros y los contras (si hubieran) de considerar $\theta = 1, 0'5, 0$**

Como bien se ha visto en clase (por tanto no ahondo demasiado en ello) la idea es aplicar el _theta method_ tal y como se ha desarrollado, es decir, nuestro objetivo es, en lugar de resolver:

<center>$\partial_tU(x,t)|_{t = t_{\theta}} - D\partial^2_{xx}U(x,t_{\theta}) = f(x,t_{\theta})$</center>
<br>

Se va a resolver (ya tomamos $D = 1$ pues es como se va a trabajar de aquí en adelante):

<center>$\frac{u_i^{j+1} - u_i^j}{\Delta t} = \theta\frac{u_{i+1}^{j+1} - 2u_i^{j+1} + u_{i-1}^{j+1}}{\Delta x^2} + (1-\theta)\frac{u_{i+1}^j - 2u_i^j + u_{i-1}^j}{\Delta x^2} + \theta f_i^{j+1} + (1-\theta)f_i^j$</center>
<br>

Siguiendo con esto (todo esto se ha visto en teoría) y tomando $\lambda := \Delta t/\Delta x^2$ se obtiene el siguiente _esquema theta_:

<center>$(1+2\theta\lambda)u_i^{j+1} - \lambda\theta(u_{i+1}^{j+1} + u_{i-1}^{j+1}) = \lambda(1-\theta)(u_{i+1}^j + u_{i-1}^j) + u_i^j(1-2\lambda(1-\theta)) + \Delta t(\theta f_i^{j+1} + (1-\theta)f_i^j) $</center>

<br>
El siguiente paso en la implementación del método sería escribir esto mismo en forma de matricial, pero como se pide en el siguiente apartado, se hace ahí. Dicho lo cual, solo queda discutir la elección del parámetro theta y comentar ventajas y desventajas de tomar $\theta = 0$, $\theta = 0'5$ o $\theta = 1$.

En cuanto a esto, un primer comentario y que también se ha visto previamente en las clases de teoría es que, la elección de $\theta = 0'5$ (método de Cranck-Nicolson) supone cometer el menor error entre las posibles elecciones de theta (comentándolo por encima, ya vimos que $T_h = \Theta(\Delta t, \Delta x^2)$ para $\theta \neq 1/2$ y $T_h = \Theta(\Delta t^2, \Delta x^2)$ para $\theta = 1/2,$ donde se usa que, en nuestro caso, el mallado que se va a construir es de tal forma que $\Delta t << 1$, luego $\Delta t^2 < \Delta t$)

Por otro lado, siguiendo otra vez los apuntes de _theta method_, ya vimos que una de las condiciones que se tienen que cumplir para que nuestro método numérico sea estable y convergente es que $\lambda(1-\theta) \leq 1/2$. Para $\theta = 1$ esa condición se cumple siempre, luego puede ser una de las ventajas de esa elección el hecho de que no va a haber ninguna restricción a la hora de "hacer" nuestro mallado y por tanto tomar un valor de $\lambda$ cualquiera. Sin embargo, para $\theta = 1/2$ se deduce que $\lambda \leq 1$ y, para $\theta = 0$, $\lambda \leq 1/2$, con lo que tenemos dichas restricciones a la hora de hacer nuestro mallado, es decir, a la hora de tomar $\Delta x$ y $\Delta t$.

Desde mi punto de vista la discusión de qué valor de $\theta$ puede ser más correcto está entre esos dos aspectos, por un lado el error cometido en los distintos casos y, por otro, las restricciones que podemos encontrar a la hora de "crear" nuestro mallado. En primer lugar, me parece razonable descartar el valor de $\theta = 0$ ya que, estando en desventaja frente a $\theta = 1/2$ si nos referimos al error, además restringe más a la hora de realizar el mallado que con $\theta = 1/2$.

En cuanto a las ventajas y desventajas que pueden tener $\theta = 1$ y $\theta = 1/2$, a priori parece razonable pensar que el segundo valor puede ser preferible en tanto y en cuanto mejora el error del método y es el que yo elegiría para nuestro caso. 

Sin embargo, también hay que comentar una cosa y es la eficiencia computacional de nuestro método. Eligiendo el construir un mallado muy fino (algo razonable pues cuanto más cerca del continuo sabemos que la solución se comportará mejor), esto supone que, cuanto más fino el mallado espacial (más pequeño $\Delta x$), $\Delta t$ tiene que ser notablemente más pequeño por la restricción que se tiene en el valor de $\lambda$ y por tanto el ordenador va a tener que realizar muchos cálculos y el método puede llegar a demorarse bastante. Por otro lado, elegir el valor $\theta = 1$ nos quita este problema y es que al no tener una restricción en el mallado, podemos hacer que el número de cálculos que realice el ordenador sea menor (aún así serían muchos pues se busca un mallado fino, pero no tantos al no tener una relación que cumplir entre $\Delta x$ y $\Delta t$) y, si el error cometido es pequeño, puede ser que con este nos sintiéramos cómodos y decidiéramos usar este valor de $\theta$ por elegir la eficiencia computacional.

Dicho lo cual, no creo que en nuestro caso, que es un caso sencillo, hubiera que preocuparse demasiado por el tiempo de demora del código, así que creo que, con vistas a reducir al máximo el error, se debería usar $\theta = 1/2$ aunque sin perder de vista que, en casos más complejos, quizá esto no es lo ideal.

#### **g) Deduce cómo implementar correctamente las condiciones frontera y escribe el esquema en forma matricial con la condición frontera ya implementada.**

Como nos encontramos en condiciones de frontera de tipo Neumann homogéneas (es el problema que hemos dicho en el apartado anterior que queremos resolver al menos de momento), veamos cómo implementar estas y cómo queda el esquema en forma matricial.

Para ello, en primer lugar fijémonos en el _esquema theta_ que se ha enseñado en el apartado anterior. Lo que ocurre es algo que ya discutimos en clase de teoría y es que, para hallar los valores en la frontera, es decir, en $i = 0, N$, se necesitan los puntos en $i = -1, N+1$ respectivamente (lo que llamábamos puntos fantasma), entonces, hay que ver cómo podemos hallar dichos puntos de cara a resolver nuestro problema numérico.

Una vez más en esta asignatura, este problema se resuelve usando la derivada centrada en dichos extremos:

<center>$(u_0^j)' = \frac{u_1^j - u_{-1}^j}{2\Delta x}$</center>
<br>
<center>$(u_N^j)' = \frac{u_{N+1}^j - u_{N-1}^j}{2\Delta x}$</center>
<br>

Ahora bien, las condiciones Neumann son homogéneas, con lo cual, $(u_0^j)', (u_N^j)' = 0$, de donde:

#### **i) Implementa el método para solucionar el problema y representa la solución obtenida**

Lo que se va tratar de hacer es un código lo más general posible y que no dependa a priori de una elección de $\theta$ o $\lambda$ (equivalentemente del número de puntos que consideremos en nuestro mallado), ya que, realmente la única ventaja de decidir qué valores usar antes de tener el código es que la matriz $A$ estaría ya determinada sin problemas, pero tampoco supone un mayor problema:

Antes, una observación para entender lo que se hace. Ya vimos que $\lambda = \Delta t /\Delta x^2$ y teniendo en cuenta que $\Delta x = 1/N$ y $\Delta t = T/M$, se tiene que se puede escribir $\lambda$ de la siguiente manera (de forma aún más general podríamos suponer que el espacio no es entre 0 y 1 si no entre 0 y L. Lo comento, pero no sigo con ello por no ser relevante):

<center>$\lambda = \frac{N^2T}{M}$</center>
<br>

Pongo en una única celda todas las funciones que podamos necesitar para nuestro objetivo con el máximo detalle posible para después simplemente llamarlas con los "números" que queramos tras haber reflexionado sobre esto.

```python
#Función que nos genera la matriz A. Es una función muy muy sencilla.

def Construye_A(theta,T,N,M): #Igual que se muestra en el enunciado N+1 es el número de puntos del mallado espacial y M+1 temporal
    _lambda = N**2*T/M
    A = np.diag((1+2*_lambda*theta)*np.ones(N+1),0) + np.diag(-_lambda*theta*np.ones(N),1) + np.diag(-_lambda*theta*np.ones(N),-1)
    #La construcción de A es bastante trivial usando np.diag()
    #Solo queda modificar las entradas con el 2
    A[0,1] = 2*A[0,1]
    A[-1,-2] = 2*A[-1,-2]
    return(A)


#Función que nos genera el array b de datos conocidos.

#Un apunte importante que quiero hacer sobre esta función es que la idea es que tome como input la fila de u previamente calculada (conocida) y...
#... por otro lado, ha de tomar también el término de producción f, pero necesita tanto en el instante en el que estemos (fila actual) como el...
#... siguiente y eso añade dificultad al código pues no todo está en un mismo instante (misma fila), así que aquí, para cada cálculo del vector b...
#... hay que tomar la f asociada a ese instante y también la siguiente, es decir, será una matriz de 2 filas.


def Construye_b(theta,T,N,M,u,f): #Mismos argumentos que en la construcción de A pues son las características de nuestro mallado y...
    #...también toma arrays u,f que son el array de calor y matriz de producción para cada instante temporal (todo conocido)
    _lambda = N**2*T/M
    
    extremos_b = [] #Los extremos del array de b se construyen con los puntos fantasma, así que lo hago fuera del bucle principal.
    extremos_b.append(_lambda*(1-theta)*2*u[1] + u[0]*(1-2*_lambda*(1-theta)) + T/M*theta*f[1,0] + T/M*(1-theta)*f[0,0]) #Extremo 0
    extremos_b.append(_lambda*(1-theta)*2*u[-2] + u[-1]*(1-2*_lambda*(1-theta)) + T/M*theta*f[1,-1] + T/M*(1-theta)*f[0,-1]) #Extremo 1
    
    b = [] #Lista vacía para el array de b (sin considerar extremos)
    for i in range(1,len(u)-1):
        b.append(_lambda*(1-theta)*(u[i-1]+u[i+1]) + u[i]*(1-2*_lambda*(1-theta)) + T/M*theta*f[1,i] + T/M*(1-theta)*f[0,i])
    
    return(np.concatenate(([extremos_b[0]],b,[extremos_b[1]]))) #Al ser números hay que convertirlos a array para concat


#Función que nos genera la matriz de calor.

def MatrizCalor(theta,T,N,M,u0,f): #Mismos argumentos, pero ahora u0 es el dato inicial conocido y f la matriz de producción (entera) conocida
                                    #f estará previamente construida "a mano" y si nos damos cuenta va a marcar hasta dónde llega el bucle
    _lambda = N**2*T/M
    HeatMatrix = np.array([u0]) #Iniciamos con el dato inicial como es evidente y usamos el formato de matriz, teniendo que u0 es un array
    A = Construye_A(theta,T,N,M) #No cambia en cada iteración con lo que se saca del bucle
    for j in range(np.shape(f)[0]-1): #El bucle llega hasta que se acaba el tiempo que estemos considerando, es decir, hasta que acabe f
        b = Construye_b(theta,T,N,M,HeatMatrix[j,:],f[[j,j+1],:]) #Como dijimos antes, esa función toma solo las dos filas de f necesarias
        HeatMatrix = np.concatenate((HeatMatrix, [np.linalg.solve(A,b)])) #Démonos cuenta que el calor es, como se dijo, el vector solución de ese sist
    return(HeatMatrix)


#Función para "meter" el dato inicial y término de producción de forma general

def Construye_u0_f(T,N,M): #Solo necesitan la información del mallado
    global x,t
    x = np.linspace(0,1,N+1)
    t = np.linspace(0,T,M+1)
    
    u0 = x**2*(1-x)**2
    f = np.zeros((len(t),len(x))) #Inicializamos ahí y ahora se rellena con un bucle, tiene la ventaja de que para t>pi ya está hecho
    
    j = 0 #Voy a hacer un while empezando en 0
    while t[j]<=np.pi:
        f[j,:] = (1+np.cos(t[j]))*x*(1-x)
        j = j+1
    return u0, f
```

```python
theta = 1
T = 5
M = int(1000)
N = int(1000)

u0, f = Construye_u0_f(T,N,M)
Calor = MatrizCalor(theta,T,N,M,u0,f)

#Un inciso es que, como declaro las variables globales x y t en la función, cada vez que la llame va a sacar un x y un t que depende del mallado...
#... y como pretendo cambiar el mallado a continuación (lo explico después) esto no es útil porque en cada llamada cambia x y t con lo que las...
#... gráficas me pueden dar problemas. Esto se soluciona recuperando dichas variables globales y guardándolas en otras:

x_theta1 = x
t_theta1 = t

#Ploteo:
plt.pcolormesh(x_theta1,t_theta1,Calor, cmap = 'inferno')
plt.title('Calor en el caso de theta = 1')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar()
plt.show()

```

#### **k) Repite el apartado i) para las condiciones Dirichlet $u(0,t) = u(1,t) = 0$. Comprueba, tal y como se dedujo teóricamente, que la solución con las condiciones homogéneas Dirichlet se encuentra por debajo de la solución con las condiciones homogéneas Neumann. Modificando el dato inicial (y, si es necesario, poniendo $f = 0$) muestra que si el dato inicial no es positivo entonces no necesariamente se tiene que la solución Neumann mayora a la de Dirichlet. ¿Puedes dar una explicación física a este hecho?**

Para este apartado no profundizo tanto en la explicación del código. Además, simplemente voy a reutilizar lo que ya teníamos antes para las condiciones Neumann ya que, esencialmente, el código es el mismo salvando un par de detalles, así que no merece la pena darle más vueltas.

```python
#Nueva función para construir A:

def Construye_A_Dirichlet(theta,T,N,M): #Igual que se muestra en el enunciado N+1 es el número de puntos del mallado espacial y M+1 temporal
    _lambda = N**2*T/M
    A = np.diag((1+2*_lambda*theta)*np.ones(N-1),0) + np.diag(-_lambda*theta*np.ones(N-2),1) + np.diag(-_lambda*theta*np.ones(N-2),-1)
    #He cambiado las dimensiones de la matriz A
    return(A)

#Nueva función para construir b:

def Construye_b_Dirichlet(theta,T,N,M,u,f): #Obsérvese que simplemente cambia el "tocar los extremos". Ahora supongo que lo que va a comer "u"...
                                            #... es simplemente la fila del calor quitando extremos, con lo que no hay que diferenciar nada
    _lambda = N**2*T/M
    
    b = [] 
    for i in range(1,len(u)-1):
        b.append(_lambda*(1-theta)*(u[i-1]+u[i+1]) + u[i]*(1-2*_lambda*(1-theta)) + T/M*theta*f[1,i] + T/M*(1-theta)*f[0,i])
    
    return(b) #Es exactamente la función anterior quitando lo de los extremos

#Función para la matriz del calor Dirichlet. Nótese la similitud con la anterior. Comento los cambios a medida que salen:

def MatrizCalorDirichlet(theta,T,N,M,u0,f): #f y u0 no cambian
    _lambda = N**2*T/M
    HeatMatrix = np.array([u0]) 
    A = Construye_A_Dirichlet(theta,T,N,M) 
    
    for j in range(np.shape(f)[0]-1): 
        b = Construye_b_Dirichlet(theta,T,N,M,HeatMatrix[j,:],f[[j,j+1],:]) #Quito extremos pues ya son conocidos por las condiciones Dirichlet
        HeatMatrix = np.concatenate((HeatMatrix, [np.concatenate(([0], np.linalg.solve(A,b), [0]))])) #Recupero concatenando los extremos nulos
        
    return(HeatMatrix)
```

Ahora llamo a la función con los mismos parámetros que concluí que podían ser mejores:

```python
theta3 = 1/2
T = 5
M3 = int(8000)
N3 = int(40)
#Estos valores cumplen la condición pues 40^2*5 = 8000, que es el caso límite en el que lambda vale 1

u03, f3 = Construye_u0_f(T,N3,M3)
Calor3 = MatrizCalorDirichlet(theta3,T,N3,M3,u03,f3)

#Mismo razonamiento. Realmente el mallado es el mismo que en el caso 2, pero hago explícito el cambio
x_theta3 = x
t_theta3 = t


plt.pcolormesh(x_theta3,t_theta3,Calor3, cmap = 'inferno')
plt.title('Calor en el caso theta = 1/2')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar()
plt.show()
```
