## **APARTADO 2. PROBLEMAS MAL PUESTOS**

#### **Considera el siguiente problema de evolución:**

**\begin{array}{rcl}
     \frac{\partial u}{\partial t}(x,y,t) & = & 2\frac{\partial^2u}{\partial x^2} - \frac{\partial^2u}{\partial y^2} : t>0, x,y\in (0,1)
  \\ u(x,y,0) & = & sin(\pi x)sin(\pi y) : t>0, x,y \in (0,1)
  \\ u(x,0,t) & = & u(x,1,t) = u(0,y,t) = u(1,y,t) = 0 : t>0, x,y \in (0,1)
\end{array}**

#### **donde $u(x,y,t)$ está definida en $x,y \in [0,1]$ y $t \geq 0$, tiene dos derivadas en $x$ e $y$ y una derivada en $t$.**

#### **a) Encuentra $a \in \mathbb R$ para que $u(x,y,t) = e^{at}sin(\pi x)sin(\pi y) : x,y \in (0,1), t>0$ sea solución de lo anterior**

En esta ocasión basta introducir $u(x,y,t)$ con esa forma en el problema de contorno y ver para qué $a$ se obtiene la solución.

Empecemos comprobando que cumple las condiciones iniciales y de contorno:

$u(x,y,0) = sin(\pi x)sin(\pi y)$ $\forall a \in \mathbb R$ de forma trivial.

$u(x,0,t) = u(x,1,t) = u(0,y,t) = u(1,y,t) = 0$ $\forall a \in \mathbb R$ también sin más que comprobarlo.

Ahora comprobemos para qué valor de $a$ dicha función es solución de la EDP:

\begin{array}{rcl}
     \frac{\partial u }{\partial t}(x,y,t) & = & au(x,y,t)
  \\ \frac{\partial^2 u}{\partial x^2} & = & -\pi^2u(x,y,t)
  \\ \frac{\partial^2 u}{\partial y^2} & = & -\pi^2u(x,y,t)
\end{array}

Sin más que igualar se deduce rápidamente que $a = -\pi^2$ para que dicha función sea solución del problema que se tiene.

#### **b) Demuestra que el problema (1) no está bien puesto ya que no hay dependencia continua del dato inicial. Para ello, dado $T > 0$ y $\epsilon > 0$, encuentra $f \in C^{\inf}((0,1)^2)$ tal que $|f(x,y)| < \epsilon $ para todo $x,y \in (0,1) $ y la solución del problema:**

**\begin{array}{rcl}
     \frac{\partial v}{\partial t}(x,y,t) & = & 2\frac{\partial^2v}{\partial x^2} - \frac{\partial^2v}{\partial y^2} : t>0, x,y\in (0,1)
  \\ v(x,y,0) & = & sin(\pi x)sin(\pi y) + f(x,y) : t>0, x,y \in (0,1)
  \\ v(x,0,t) & = & v(x,1,t) = u(0,y,t) = u(1,y,t) = 0 : t>0, x,y \in (0,1)
\end{array}**

#### **cumple que $sup\{|v(x,y,T) - u(x,y,T)|: x,y \in (0,1)^2\} > 1$.**

Como la EDP que cumple $v(x,y,t)$ es exactamente la misma que en el caso anterior, nos ayudamos de lo deducido en el apartado a).

Aunque se podría tratar de hacer de un modo general este ejercicio, voy a tratar de explicar el procedimiento que he llevado a cabo para encontrar dicha $f(x,y)$ de forma más rápida. En primer lugar, se ve "a ojo" que la función $v(x,y,t) = e^{-\pi^2t}sin(\pi x)sin(\pi y) + C(t)f(x,y)$ es solución de este problema siempre y cuando $C(0) = 1$ ya que el primer sumando sabemos que lo es por el apartado anterior. Además, obviamente, $C(t)f(x,y)$ debe ser solución de la EDP y cumplir las condiciones de contorno con lo que simplemente he pensado en una solución que tenga una forma muy similar a $e^{-\pi^2t}sin(\pi x)sin(\pi y)$. Se sabe que, de forma teórica, en las soluciones a este tipo de EDP's aparecen series de senos y cosenos y, dado que las condiciones de contorno me dicen que $C(t)f(x,0) = C(t)f(x,1) = C(t)f(0,y) = C(t)f(1,y) = 0$ se deduce que:

$C_n(t)f_n(x,y) = e^{2n^2\pi ^2t}sin(n\pi x)sin(2n\pi y) \forall n \in \mathbb N$ cumplen con lo que se ha buscado hasta el momento de forma trivial. El buscar $C_n(t)$ con esa forma exponencial se debe a que he buscado una función similar a la anterior y que se requería que $C(0) = 1$

Así, $v_n(x,y,t) = e^{-\pi^2t}sin(\pi x)sin(\pi y) + e^{2n^2\pi ^2t}sin(n\pi x)sin(2n\pi y)$ son soluciones de este problema.

Una vez más, incidir en que se ha llegado a esta expresión sin más que usar el apartado a) y ver lo que cumplía la función $f(x,y)$ para deducir que se podría dar una para cada $n$ natural con la misma estructura que la solución hallada antes.

Por otro lado, se pide que la perturbación $f(x,y)$ sea pequeña, esto es, dado $\epsilon > 0$ arbitrario $|f(x,y)| < \epsilon$, para ello se puede introducir, como se suele hacer, un peso a las funciones $f_n(x,y)$:

$f_n(x,y) = \frac{\epsilon}{2}sin(n\pi x)sin(2n\pi y)$. Dado que los senos están acotados por 1, se cumple lo que se pide.

De esta forma, además, la exponencial de $C_n(t)$ va a ser el motivo que haga que se cumpla la condición del supremo que se busca ya que:

$|v_n(x,y,t) - u(x,y,t)| = e^{2n^2\pi ^2t}|sin(n\pi x)sin(2n\pi y)|$

De esta forma ya es claro que, gracias a la exponencial como se adelantó antes, se puede escoger una $n$ para que el supremo de dicho conjunto sea todo lo grande que se quiera, es decir, se puede tomar tal que $sup\{|v(x,y,T) - u(x,y,T)|: x,y \in (0,1)^2\} > 1$ que es lo que se buscaba y, por tanto, este problema está mal puesto en tanto y en cuanto no hay dependencia continua del dato inicial.

## **APARTADO 3. DIFERENCIAS FINITAS**

Para hacer este ejercicio, en los dos primeros apartados se requiere de resolver dos sistemas de ecuaciones lineales. La idea es aprovechar Python para resolverlo de forma más rápida.

**Sea $f$ una función suficientemente diferenciable y con todas sus derivadas acotadas. Dado un $h>0$ se pide:**

#### **c) Considerar las diferencias finitas de la forma forward $\Delta_hf(x) = \frac{a_0f(x) + a_1f(x+h) + a_2f(x+2h) + a_3f(x+3h)}{h}$:**

#### **Determinar los valores $a_0$, $a_1$, $a_2$, $a_4 \in \mathbb R$ que hagan que $\Delta_hf(x)$ sea una discretización de $f'(x)$ del mayor orden posible.**

Para la resolución de este ejercicio basta aplicar el razonamiento visto en clase de teoría y aplicar desarrollos en serie de Taylor para los términos $f(x+h)$, $f(x+2h)$ y $f(x+3h)$ cuando $h$ tiende a 0. Para ello, se trunca en el orden 3 (no se puede avanzar más para lo que se busca hacer, se explica más adelante). De esta forma:

\begin{array}{rcl}
     f(x+h) & = & f(x) + hf'(x) + \frac{1}{2}h^2f''(x) + \frac{1}{6}h^3f'''(x) + O(h^4)
  \\ f(x+2h) & = & f(x) + 2hf'(x) + 2h^2f''(x) + \frac{4}{3}h^3f'''(x) + O(h^4)
  \\ f(x+3h) & = & f(x) + 3hf'(x) + \frac{9}{2}h^2f''(x) + \frac{9}{2}h^3f'''(x) + O(h^4)
\end{array}

Ahora estamos en condiciones de sustituir por la expresión del enunciado con lo que se llega a: 
$\Delta_hf(x) = (a_0+a_1+a_2+a_3)\frac{f(x)}{h} + (a_1+2a_2+3a_3)f'(x) + (\frac{1}{2}a_1+2a_2+\frac{9}{2}a_3)hf''(x) + (\frac{1}{6}a_1+\frac{4}{3}a_2+\frac{9}{2}a_3)h^2f'''(x)$

Ahora, téngase en cuenta que se busca aproximar $f'(x)$, luego está claro que se han de cumplir las siguientes condiciones:
\begin{array}{rcl}
     a_0 + a_1 + a_2 + a_3 & = & 0
  \\ a_1 + 2a_2 + 3a_3 & = & 1
  \\ \frac{1}{2}a_1+2a_2+\frac{9}{2}a_3 & = & 0
  \\ \frac{1}{6}a_1+\frac{4}{3}a_2+\frac{9}{2}a_3 & = & 0
\end{array}

Queda por tanto, resolver ese sistema para hallar los coeficientes $a_j$ que cumplen lo que buscamos, que la discretización de $f'(x)$ sea del mayor orden posible. Para ello usamos np.linalg.solve()

---

```python
A1 = np.array([[1,1,1,1],[0,1,2,3],[0,1/2,2,9/2],[0,1/6,4/3,27/6]])
B1 = np.array([0,1,0,0])

solucion1 = np.linalg.solve(A1,B1)
solucion1
```

Aunque numpy muestra los números con decimales, podemos pasarlo a fracciones y se obtiene que los coeficientes en cuestión son:

$a_0 = -\frac{11}{6}$, $a_1 = 3$, $a_2 = -\frac{3}{2}$ y  $a_3 = \frac{1}{3}$

#### **d) Considerar las diferencias finitas centradas de la forma $\delta_hf(x) = \frac{b_{-2}f(x-2h) + b_{-1}f(x-h) + b_1f(x+h) + b_2f(x+2h)}{h}$:**

#### **Determinar los valores $b_{-2}$, $b_{-1}$, $b_1$, $b_2 \in \mathbb R$ que hagan que $\delta_hf(x)$ sea una discretización de $f'(x)$ del mayor orden posible.**

De nuevo el razonamiento es el mismo que en el apartado anterior aunque ahora tengamos diferencias finitas centradas, el resto es totalmente análogo. Desarrollando por Taylor los distintos elementos se tiene:

\begin{array}{rcl}
     f(x-2h) & = & f(x) - 2hf'(x) + 2h^2f''(x) - \frac{4}{3}h^3f'''(x) + O(h^4)
  \\ f(x-h) & = & f(x) - hf'(x) + \frac{1}{2}h^2f''(x) - \frac{1}{6}h^3f'''(x) + O(h^4)
  \\ f(x+h) & = & f(x) + hf'(x) + \frac{1}{2}h^2f''(x) + \frac{1}{6}h^3f'''(x) + O(h^4)
  \\  f(x+2h) & = & f(x) + 2hf'(x) + 2h^2f''(x) + \frac{4}{3}h^3f'''(x) + O(h^4)
\end{array}

Sin más que sustituir en la expresión de enunciado, se tiene que:
$\delta_hf(x) = (b_{-2} + b_{-1} + b_1 + b_2)\frac{f(x)}{h} + (-2b_{-2} - b_{-1} + b_1 + 2b_2)f'(x) + (2b_{-2} + \frac{1}{2}b_{-1} + \frac{1}{2}b_1 + 2b_2)hf''(x) + (-\frac{4}{3}b_{-2} - \frac{1}{6}b_{-1} + \frac{1}{6}b_1 + \frac{4}{3}b_2)h^2f'''(x)$

De nuevo, todo se reduce a resolver el siguiente sistema:
\begin{array}{rcl}
     b_{-2} + b_{-1} + b_1 + b_2 & = & 0
  \\ -2b_{-2} - b_{-1} + b_1 + 2b_2 & = & 1
  \\ 2b_{-2} + \frac{1}{2}b_{-1} + \frac{1}{2}b_1 + 2b_2 & = & 0
  \\ -\frac{4}{3}b_{-2} - \frac{1}{6}b_{-1} + \frac{1}{6}b_1 + \frac{4}{3}b_2 & = & 0
\end{array}

```python
A2 = np.array([[1,1,1,1],[-2,-1,1,2],[2,1/2,1/2,2],[-4/3,-1/6,1/6,4/3]])
B2 = np.array([0,1,0,0])

solucion2 = np.linalg.solve(A2,B2)
solucion2
```

De nuevo, en forma de fracción para mayor comodidad:

$b_{-2} = \frac{1}{12}$, $b_{-1} = -\frac{2}{3}$, $b_1 = \frac{2}{3}$, $b_2 = -\frac{1}{12}$

Un comentario (no incido mucho en ello) que se quiere hacer sobre estos dos apartados es que se ha buscado desde el primer momento discretizar con el mayor orden posible dado que se pide en el enunciado.

En el caso del método forward, tomar un orden más en el desarrollo de Taylor y razonando igual, hace que el sistema que queda a resolver sea incompatible y que, por tanto, no se pueda discretizar con más de orden del que se ha mostrado, por tanto, el máximo es orden 3. No se muestra por carecer de interés y porque en el caso de la derivada centrada se va a exponer que es lo que pasa en este caso también.

Sin embargo, para el caso del método centrado, la ecuación que se añade al sistema debido a tomar un orden más en el desarrollo de Taylor no afecta al sistema, es decir, se siguen obteniendo las mismas soluciones para las $b_i$ y, por tanto, el orden máximo en este caso es 4 (aumentando más el sistema ya no tiene solución).

Siendo más concretos, si se considera un orden más en el desarrollo de Taylor, a $f(x\pm 2h)$ hay que sumarle $\frac{2}{3}h^4f^{(4)}(x)$ y a $f(x\pm h)$ se le suma $\frac{1}{24}h^4f^{(4)}(x)$ con lo que, razonando exactamente igual que antes, se llega a que para que la aproximación de $f'(x)$ sea lo mejor posible, se tiene que resolver el siguiente sistema:

\begin{array}{rcl}
     b_{-2} + b_{-1} + b_1 + b_2 & = & 0
  \\ -2b_{-2} - b_{-1} + b_1 + 2b_2 & = & 1
  \\ 2b_{-2} + \frac{1}{2}b_{-1} + \frac{1}{2}b_1 + 2b_2 & = & 0
  \\ -\frac{4}{3}b_{-2} - \frac{1}{6}b_{-1} + \frac{1}{6}b_1 + \frac{4}{3}b_2 & = & 0
  \\ \frac{2}{3}b_{-2} + \frac{1}{24}b_{-1} + \frac{1}{24}b_1 + \frac{2}{3}b_2 & = & 0
\end{array}

En lugar de resolver el sistema (np.linalg.solve toma matrices cuadradas como input) veo que los valores hallados antes cumplen esa última ecuación:

```python
np.around(2/3*solucion2[0] + 1/24*solucion2[1] + 1/24*solucion2[2] + 2/3*solucion2[3]) #Redondeo para que no haya problemas numéricos de Python
```

#### **e) Considerad $f(x) = cos(x)$. Construid un programa y calculad numéricamente $f'(x)$ en el intervalo $x \in [0,2\pi]$ utilizando las discretizaciones $\Delta_h$, $\delta_h$ que acabais de obtener. Calculad el error de vuestra derivada en ambos casos para diferentes valores de $h$ y discutid si el error se ajusta a las predicciones teóricas. Recomendación: pintar el error en escala logarítmica.**

##### Primer caso: $\Delta_h$:
Démonos cuenta que, a priori, parecería un problema el hecho de usar el método forward para calcular numéricamente la derivada en los últimos puntos del dominio (nosotros vamos a tener un array en el cual queremos calcular derivadas) ya que para llegará un momento (últimos 3 puntos) en el que nos saldremos del array.
Para solucionar este problema, se usa que la función $cos(x)$ es una función 2$\pi$-periódica con lo que se puede tomar el inicio del array como "continuación" del propio array del dominio ya que estamos en $[0,2\pi]$

```python
def DerivadaForward(solucion1,numPuntos): #solucion1 van a ser los hallados antes en el apartado c, se hace así para aportar generalidad
    #numPuntos va a ser el número de puntos que consideramos en el dominio, lo ponemos como input de cara a variarlo más adelante
    t=np.linspace(0,2*np.pi,numPuntos) #Dominio que se tiene en cuenta
    
    t_modificado=np.concatenate((t,np.array([t[1],t[2],t[3]]))) #Como se ha explicado, de esta forma vamos a poder calcular la derivada forward...
    #... de los 3 últimos elementos del array sin salirnos del mismo en el bucle que viene a continuación. Otra forma sería recorrer t con el...
    #... for y poner condiciones al llegar a los 3 últimos elementos, pero así el código queda más sencillo.
    f=np.cos(t_modificado) #Queremos la imagen de t_modificado para usar esa lógica
    global h #Se hace esto para que la variable h se pueda usar fuera de la función, que nos será útil para después, pero no la saco como...
    #...Output ya que no tiene mayor relevancia
    h=t[1]-t[0] #Como es un linspace, el paso es siempre el mismo y uso para ellos los dos primeros elementos del dominio (su diferencia)
    derivada_forward=[]
    for i in range (0,len(t_modificado)-3):
        derivada_forward.append((solucion1[0]*f[i]+solucion1[1]*f[i+1]+solucion1[2]*f[i+2]+solucion1[3]*f[i+3])/h) 
    return(t, derivada_forward)
```

```python
t, derivada_forward = DerivadaForward(solucion1,100)
plt.plot(t,derivada_forward, label = 'Derivada (método forward)')
plt.legend()
plt.xlabel('t')
plt.ylabel('$\Delta_h f$')
plt.title('Derivada forward')
plt.show()
```

##### Segundo caso $\delta_h$:

De nuevo puede parecer que tenemos el mismo problema con el tema de "salirse" del dominio, pero se soluciona de la misma manera aunque ahora también hay que añadir los elementos finales al principio del array para obtener los términos $f(x-2h)$ y $f(x-h)$.

```python
def DerivadaCentrada(solucion2,numPuntos): #solucion2 van a ser los hallados antes en el apartado d, se hace así para aportar generalidad
    t=np.linspace(0,2*np.pi,numPuntos) #Dominio que se tiene en cuenta
    
    t_modificado=np.concatenate((np.array([t[-3],t[-2]]),t,np.array([t[1],t[2]]))) #No se le da más vueltas, misma lógica
    f=np.cos(t_modificado)
    global h
    h=t[1]-t[0] #Otra vez, mismo salto siempre 
    derivada_centrada=[]
    for i in range (2,len(t_modificado)-2):
        derivada_centrada.append((solucion2[0]*f[i-2]+solucion2[1]*f[i-1]+solucion2[2]*f[i+1]+solucion2[3]*f[i+2])/h)
    return(t,derivada_centrada)
```

```python
t, derivada_centrada = DerivadaCentrada(solucion2,100)
plt.plot(t,derivada_centrada, label = 'Derivada (método centrado)')
plt.legend()
plt.xlabel('t')
plt.ylabel('$\delta_h f$')
plt.title('Derivada Centrada')
plt.show()
```

##### Error entre las discretizaciones y la derivada:

Basta llamar a las funciones con las que hemos calculado las derivadas forward y centradas con distinto cardinal del dominio o, lo que es lo mismo, a más puntos en el dominio como se crea un linspace, un menor paso $h$.

La idea es que para cada valor de $h$ se puede calcular un error asociado a cada punto del linspace, lo cual nos da un vector de errores para cada valor de $h$. Como el objetivo es representar el error para cada valor de $h$, parece lógico que, para cada vector de errores, calcular la media y así, para cada $h$ tendremos un único error.

```python
cardinal_lista = np.linspace(10,1000,100,dtype=int) #Por ejemplo, se podrían tomar todos los que se quisieran
h_lista = [] #Lista vacía con las h sucesivas

#Listas con las medias de errores vacías
error_forward_mean = []
error_centrada_mean = []

for i in range(0,len(cardinal_lista)):    
    #Ahora llamamos a las funciones para obtener las derivadas forward y centrada para cada valor del cardinal del dominio
    t, derivada_forward = DerivadaForward(solucion1,cardinal_lista[i])
    t, derivada_centrada = DerivadaCentrada(solucion2,cardinal_lista[i])
    h_lista.append(h) #Como la variable h es global, se le puede llamar sin problema
    f_der = -np.sin(t)
    error_forward_lista = np.abs(f_der - derivada_forward)
    error_centrada_lista = np.abs(f_der - derivada_centrada)
    
    error_forward_mean.append(np.mean(error_forward_lista))
    error_centrada_mean.append(np.mean(error_centrada_lista))
    
    
h_lista_log10 = np.log10(h_lista)#Tomo logaritmos en base 10 siguiendo las instrucciones del enunciado
error_FM_log10 = np.log10(error_forward_mean)
error_CM_log10 = np.log10(error_centrada_mean)

plt.scatter(h_lista_log10,error_FM_log10, color = 'red', label = 'Error Forward', s = 3.0)
plt.scatter(h_lista_log10,error_CM_log10, color = 'blue', label = 'Error Centrada', s = 3.0)
plt.xlabel('log10(h)')
plt.ylabel('log10(Mean_errores)')
plt.legend()
plt.show()
```

Ahora tiene sentido, dado que en escala logarítmica se observa una tendencia lineal de los errores con $h$ (obviamente a menor $h$ se observa un menor error con la derivada analítica como era de esperar), tiene sentido buscar la pendiente de dicha recta con lo que hallaríamos con qué exponente crece el error al aumentar $h$. Esto se puede hacer sin más que hacer un ajuste lineal:

```python
ajuste_error_forward = np.polyfit(h_lista_log10,error_FM_log10,1) #Nos devuelve pendiente y ordenada en el origen
m_forward = ajuste_error_forward[0]
n_forward = ajuste_error_forward[1]

ajuste_error_centrada = np.polyfit(h_lista_log10,error_CM_log10,1) #Igual
m_centrada = ajuste_error_centrada[0]
n_centrada = ajuste_error_centrada[1]

dominio = np.linspace(h_lista_log10[0],h_lista_log10[-1],1000)

plt.plot(dominio,m_forward*dominio + n_forward, color = 'orange', label = 'Ajuste forward')
plt.plot(dominio,m_centrada*dominio + n_centrada, color = 'green', label = 'Ajuste centrada')

plt.scatter(h_lista_log10,error_FM_log10, color = 'red', label = 'Error Forward', s = 3.0) #Vuelvo a añadir los puntos para poner el ajuste encima
plt.scatter(h_lista_log10,error_CM_log10, color = 'blue', label = 'Error Centrada', s = 3.0)
plt.xlabel('log10(h)')
plt.ylabel('log10(Mean_errores)')
plt.legend()
plt.show()

print('La pendiente del método forward es \n')
print(m_forward)
print('\n La pendiente del método centrado es \n')
print(m_centrada)
```

Se observa que las pendientes son, respectivamente, muy cercanas a 3 y 4. No parece raro pensar que, en escala logarítmica esta sea la pendiente de las rectas y que por tanto los exponentes sean estos ya que son precisamente los órdenes máximos de la discretización en ambos casos.

Teóricamente esto es lo que debería aparecer ya que los órdenes del error son, como se ha explicado antes con más detalle, propocionales a $h^3$ y $h^4$ respectivamente con lo que, tomando logaritmos a ambos lados de las ecuaciones, se tienen las pendientes de valores 3 y 4, que son precisamente lo que obtenemos ahora.

```python
np.set_printoptions(suppress=True) #Para que no aparezca notación científica de cara a redondear
warnings.filterwarnings('ignore') #Para ignorar avisos innecesarios

#Defino una función para obtener coeficientes de Fourier dado un cierto array "delta" de imágenes:
def CoefsFourier(alpha,delta):
    c_n = np.fft.rfft(delta, norm='forward') #Coeficientes en forma compleja
    #a_n y b_n serán los arrays de la representación trigonométrica de la serie de Fourier
    a_n = np.zeros(len(c_n))
    b_n = np.zeros(len(c_n)-1) #De dimensión una menos pues no tenemos b_0
    for i in range(0, len(c_n)):
        a_n[i] = c_n[i] + np.conjugate(c_n[i])
        if i>0:
            b_n[i-1] = (np.conjugate(c_n[i]) - c_n[i])/(0 + 1j)
    #Redondeo los coeficientes hasta la segunda cifra decimal. Ya trabajo todo el tiempo con los de la reprsentación trigonométrica:
    a_n = a_n.round(2)
    b_n = b_n.round(2)
    return(a_n,b_n)

#Defino una función para obtener la serie de Fourier asociada o el valor estimado en un cierto alpha:
def SerieFourier(a_n,b_n,alpha,T,dominio): #El dominio entra como input porque querremos calcular tanto arrays como valores concretos
    serie_trig = np.zeros(len(dominio))
    serie_trig = serie_trig + a_n[0]/2 #Primer valor
    for i in range(0,len(b_n)):
        serie_trig = serie_trig + a_n[i+1]*np.cos((i+1)*2*np.pi/T*dominio) + b_n[i]*np.sin((i+1)*2*np.pi/T*dominio)
    return(serie_trig)

#Defino otra función para la derivada (tasa de cambio) siguiendo el mismo argumento que en la anterior teniendo en cuenta cómo queda la derivada
def SerieFourierDer(a_n,b_n,alpha,T,dominio): #El dominio entra como input porque querremos calcular tanto arrays como valores concretos
    serie_trig_der = np.zeros(len(dominio))
    for i in range(0,len(b_n)):
        serie_trig_der = serie_trig_der - (i+1)*2*np.pi/T*a_n[i+1]*np.sin((i+1)*2*np.pi/T*dominio) + (i+1)*2*np.pi/T*b_n[i]*np.cos((i+1)*2*np.pi/T*dominio)
    return(serie_trig_der)
```

```python
#Meto los datos que tenemos de cara a resolver el ejercicio concreto que tenemos
alpha = np.arange(0,360,30)
delta = np.array([408, 89, -66, 10, 338, 807, 1238, 1511, 1583, 1462, 1183, 804])

T = 360 #Doy ya también el período en este caso que será 360 a lo largo de todo el ejercicio
```

Sin más que llamar a la función CoefsFourier de arriba se pueden obtener los coeficientes $a_n$ y $b_n$ de la serie de Fourier en su reprsentación trigonométrica:

```python
warnings.filterwarnings('ignore')
a_n,b_n = CoefsFourier(alpha,delta)
print('Los coeficientes a_n son: \n')
print(a_n)
print('\n Por otro lado, los coeficientes b_n son: \n')
print(b_n)
```

$f(\alpha) = a_0/2 + a_1cos(\frac{2\pi}{360}\alpha) + b_1sen(\frac{2\pi}{360}\alpha) + ... + a_6cos(\frac{2\pi}{360}6\alpha) + b_6sen(6\frac{2\pi}{360}\alpha)$. Más concretamente, si usamos los coeficientes encontrados más arriba:

#### **h)Representa la función $f(\alpha)$ obtenida:**

Basta llamar ahora a la función SerieFourier que toma como input los coeficientes $a_n$ y $b_n$ de la representación trigonométrica así como el array de $\alpha$ que se tenga como dato y el período T.

```python
dominio = np.linspace(alpha[0], alpha[-1], 1000) #Dominio de cara a la representación. Lo hago para que cubra los valores experimentales.

serie_trig = SerieFourier(a_n,b_n,alpha,T,dominio) #dominio es simplemente donde evaluar la suma de Taylor y...
#... serie_trig es la imagen de la suma que se ha mostrado arriba en ese dominio

plt.plot(dominio,serie_trig, color = 'green', label = 'Serie')
plt.scatter(alpha,delta, color = 'blue', label = 'Datos')
plt.legend()
plt.xlabel('alpha (rad)')
plt.ylabel('delta(min)')
plt.title('Gráfica Serie Fourier FFT')
plt.show()
```

#### **j) Finalmente, utiliza la serie de Fourier calculada para estimar la tasa de cambio de $\delta$ en función de $\alpha$ y representa la función resultante. Así mismo estima el valor de $f'(45)$. Se valorará positivamente la resolución de este ejercicio mediante el uso de un algoritmo lo más general posible.**

Para ello puedo utilizar la propia función "SerieFourierDer" teniendo en cuenta que la derivada de la suma finita de Fourier, sin más que derivar es:

$f'(t) = + 411'01sen(\alpha) - 720'23cos(\alpha) - 2\cdot43'42sen(2\alpha) - 2\cdot2'17cos(2\alpha) + 3\cdot4'33sen(3\alpha) + 3\cdot5'50cos(3\alpha) + 4\cdot1'08sen(4\alpha) - 4\cdot1'01cos(4\alpha) - 5\cdot0'35sen(5\alpha) - 5\cdot0'27cos(5\alpha) - 6\cdot0'17sen(6\alpha)$

Es decir, los coeficientes que van con los cosenos son los $n\cdot b_n$ y los que van con el seno son los $-n\cdot a_n$, que es lo que ya he tenido en cuenta en la función SerieFourierDer

Por ello, basta llamar a dicha función:

```python
dominio = np.linspace(alpha[0], alpha[-1], 1000) #Dominio de cara a la representación. Lo hago para que cubra los valores experimentales.
serie_trig_der = SerieFourierDer(a_n,b_n,alpha,T,dominio) 

plt.plot(dominio,serie_trig_der, color = 'green', label = 'Derivada')
plt.legend()
plt.xlabel('alpha (rad)')
plt.ylabel('delta(min)')
plt.title('Gráfica Derivada de la Serie de Fourier')
plt.show()
```
