## **Apartado 3: Principio del máximo**

**En clase hemos visto la existencia y unicidad de soluciones para la ecuación de Poisson, mediante los principios del máximo y el método de Perron. Sin embargo, la ecuación (1, ver enunciado original) no es la ecuación de Poisson, sino una leve modificación de ella que modeliza un medio más complejo.**

**Aún así, podemos obtener un principio del máximo de manera similar a como lo hicimos lo hicimos para la ecuación de Poisson. Vamos a probar el siguiente teorema:**

**Theorem 3.0.1: _Dado $\Omega \subset \mathbb R^d$ acotado y $u\in C^2(\Omega)\cup C^0(\bar{\Omega})$ se tiene que si $u$ satisface la inecuación:_**

<center>$L[u](x) = -\sum_{i = 1}^d\frac{\partial }{\partial x_i}\left(a_i\frac{\partial u}{\partial x_i}\right)(x) \geq 0$</center>
<br>

**_donde los coeficientes $a_i$ satisfacen (2), entonces:_**

<center>$\min_{x\in\bar{\Omega}}u(x) = \min_{x\in\partial\Omega}u(x)$</center>
<br>

**_Demostración_: Dividiremos, como en la ecuación de Poisson, la prueba en dos casos:**


**1. Caso $L[u](x) > 0$: Como las funciones $a_i$ son $C^1$, podemos expandir la inecuación a:**

<center>$-\sum_{i = 1}^d\frac{\partial a_i}{\partial x_i}(x)\frac{\partial u}{\partial x_i}(x) - \sum_{i = 1}^d a_i(x)\frac{\partial^2 u}{\partial x_i^2}(x) > 0 $  (5)</center>
<br>

#### **a) Asume que el mínimo se encuentra en el interior, es decir, que $u(x_0) = \min_{x\in\bar{\Omega}}u(x)$ con $x_0\in\Omega$. Prueba que esto lleva a una contradicción con (5):**

Por el hecho de estar trabajando con un punto interior y que se dan las hipótesis necesarias de diferenciabilidad, dado que $x_0$ es un mínimo de $u$ se tiene que:

<center>$\frac{\partial u}{\partial x_i}(x_0) = 0$</center>
<br>
<center>$\frac{\partial^2 u}{\partial x_i^2}(x_0) > 0$</center>
<br>

Así, evaluando $L[u](x)$ (de la forma que más desarrollada) en $x_0$ se deduce que:

<center>$L[u](x_0) = - \sum_{i = 1}^d a_i(x)\frac{\partial^2 u}{\partial x_i^2}(x_0) < 0$</center>
<br>

Donde se ha usado lo que se ha dicho y que, además, por hipótesis, existen $m,M > 0$ tales que $m \leq a_i(x) \leq M$, luego $a_i$ es positiva.

La hipótesis era que $L[u](x) > 0$, pero si el mínimo se encuentra en un punto interior $x_0$ se deduce que $L[u](x_0) < 0$, con lo que efecitvamente se alcanza una contradicción y el mínimo se encuentra, por tanto, en la frontera.

**2. Caso $L[u](x) \geq 0$: En este caso usamos la barrera construida en las clases de teoría $\Phi\in C^2(\bar{\Omega})$ que cumple:**

**I) $\Phi \geq 0$ , II) $L[\Phi] \leq -1$ , III) $||\Phi||_{\infty} = \max_{x\in\bar{\Omega}}\Phi(x) \leq M$**

**Usando la barrera $\Phi$:**

#### **b) Dado $\epsilon >0$, considera $v_{\epsilon} = u - \epsilon\Phi$. Prueba, usando el apartado anterior que $\min_{x\in\bar{\Omega}}v_{\epsilon}(x) =\min_{x\in\partial\Omega}v_{\epsilon}(x) $**

Como el operador $L$ solo involucra derivadas y derivadas segundas, se deduce que es un operador lineal, luego si se aplica sobre nuestra función $v_{\epsilon}$ se tiene que:

<center>$L[v_{\epsilon}](x) = L[u](x) - \epsilon L[\Phi](x)$</center>
<br>

Ahora, con esto, simplemente basta utilizar las hipótesis con "cuidado". En primer lugar se tiene que $L[u](x) \geq 0$, de donde se deduce que $L[v_{\epsilon}](x) \geq -\epsilon L[\Phi](x)$.

Por otro lado, dado que $L[\Phi](x) \leq -1$, se tiene que $L[v_{\epsilon}](x) \geq \epsilon > 0$. Estamos por tanto en la situación discutida en el apartado anterior de donde se deduce que, efectivamente:

<center>$\min_{x\in\bar{\Omega}}v_{\epsilon}(x) =\min_{x\in\partial\Omega}v_{\epsilon}(x) $</center>
<br>

#### **c) Concluye (detalladamente) que $\min_{x\in\bar{\Omega}}u(x) =\min_{x\in\partial\Omega}u(x) $**

Para este apartado, simplemente hay que utilizar los dos anteriores y lo que en ellos se ha deducido. Una primera observación sencilla, pero importante es que, dado que por hipótesis la barrera $\Phi \geq 0$, se tiene que $u(x) \geq v_{\epsilon}(x) \forall x \in \bar{\Omega}$ de donde se deduce directamente que $\min_{x\in\bar{\Omega}}u(x) \geq \min_{x\in\bar{\Omega}}v_{\epsilon}(x)$. Ahora bien, por los apartados anteriores se tiene:

<center>$\min_{x\in\bar{\Omega}}u(x) \geq \min_{x\in\bar{\Omega}}v_{\epsilon}(x) = \min_{x\in\partial\Omega}v_{\epsilon}(x) = \min_{x\in\partial\Omega}(u-\epsilon\Phi)(x) \geq \min_{x\in\partial\Omega}u(x) - \epsilon M$</center>
<br>

Donde la primera igualdad se tiene del apartado anterior, la segunda es por definición de $v_{\epsilon}$ y la última desigualdad se tiene por un lado del hecho de separar esa resta y que por tanto el mínimo de la resta sea mayor o igual que la resta de los mínimos y por otro de que $\max_{x\in\bar{\Omega}}\Phi(x) \leq M$ por hipótesis (que esencialmente da el mínimo de $-\epsilon\Phi$).

Ahora bien, se tiene que:
<center>$\min_{x\in\bar{\Omega}}u(x) \geq \min_{x\in\partial\Omega}u(x) - \epsilon M \forall \epsilon > 0$</center>
<br>
Sin más que hacer tender $\epsilon$ a $0$, se tiene que $\min_{x\in\bar{\Omega}}u(x) \geq \min_{x\in\partial\Omega}u(x)$.

La otra desigualdad es trivial en tanto y en cuanto $\partial\Omega \subset \bar{\Omega}$, luego lógicamente $\min_{x\in\bar{\Omega}}u(x) \leq \min_{x\in\partial\Omega}u(x)$.

De ambas desigualdades se deduce que, efectivamente:

<center>$\min_{x\in\bar{\Omega}}u(x) =\min_{x\in\partial\Omega}u(x) $</center>

**Corolario 1. _Sean $u_1, u_2 \in C^2(\Omega)\cup C(\bar{\Omega})$ dos soluciones del problema:_**

**<center>$L[u](x) = f(x)$ , $x\in\Omega$</center>**
**<center>$u(x) = u_0(x)$ , $x\in\partial\Omega$</center>**
<br>

**_Entonces se tiene que $u_1 = u_2$_**

**_Demostración:_**

#### **d) Prueba el corolario. Para ello, deduce del Principio del mínimo un principio del máximo y utiliza ambos para probar que $u_1 = u_2$. La prueba es similar a la usada en los resultados de parabólicas y elípticas durante las clases de teoría.**

Veamos en primer lugar que se puede deducir un principio del máximo a partir de lo que se ha deducido de _theorem 3.0.1_ (principio del mínimo). Siguiendo lo que se ha hecho en clase ya en varias ocasiones, la idea es aplicar el principio del mínimo ya probado a $-u$, pero para ello hay que tener en cuenta cómo afecta este cambio de signo a las hipótesis del sistema y es que (poniendo mismas hipótesis de diferenciabilidad) para que se dé la hipótesis de que $L[-u](x) \geq 0$, se ha de pedir que $L[u](x) \leq 0$ por la linealidad de la derivada (este cambio de desigualdad es la única modificación que hay que hacerle a las hipótesis del _theorem 3.0.1_ para conseguir nuestro principio del máximo).

Con esto sobre la mesa, dada $u$ tal que $L[u](x) \leq 0$, se tiene que, efectivamente, $L[-u](x) \geq 0$. Como $-u$ es igualmente diferenciable, aplicando el principio del mínimo se deduce que:

<center>$\min_{x\in\bar{\Omega}}(-u)(x) = \min_{x\in\partial\Omega}(-u)(x)$</center>
<br>

Ahora bien, obsérvese que $\min_{x\in A}(-u)(x) = -\max_{x\in A}(u)(x)$ (donde $A$ es un conjunto arbitrario, para no repetir dos veces lo mismo), de tal forma que, por la igualdad de arriba se tiene lo que se buscaba:

<center>$\max_{x\in\bar{\Omega}}(u)(x) = \max_{x\in\partial\Omega}(u)(x)$</center>
<br>

Una vez se tiene el análogo principio del máximo, buscamos probar el corolario en cuestión, lo cual ya es bastante sencillo, veámoslo.

Consideremos $v = u_1 - u_2$ donde $u_1, u_2$ son soluciones del problema en cuestión. Entonces, se cumple trivialmente que $L[v](x) = 0 \forall x\in\Omega$ y que $v(x) = 0 \forall x\in\partial\Omega$. Por lo primero, observemos que nos encontramos en condiciones de aplicar tanto el principio del máximo como el principio del mínimo (ya que $L[v](x) \geq 0$ y $L[v](x) \leq 0$, por dejarlo totalmente claro), luego el máximo y el mínimo de $v$ se alcanzan en $\partial\Omega$ donde la función es nula, luego se deduce que efectivamente:

<center>$v(x) \equiv 0 \Longrightarrow u_1 = u_2$</center>

## **Apartado 5: El método numérico**

**Para resolver el problema vamos a considerar un mallado de $\Omega$, el cual, al tener una forma cuadrangular admite un mallado sencillo uniforme en cada una de las direcciones del espacio. Haciendo un mallado de $N + 2$ puntos en la variable $x$ y $M+2$ puntos en la variable $y$, denominaremos $h = (h_1,h_2)$ donde $h_1 = \frac{3}{N+1}$ y $h_2 = \frac{3}{M+1}$, $\Omega_h$ al mallado interior y $\partial\Omega_h$ al mallado frontera.**

**El método numérico que nosotros vamos a utilizar es el que vimos en el curso para este tipo de ecuaciones:**

<center>$A_hu_h(x,y) = -\partial_x(a(\cdot -\frac{h_1}{2}\vec{e_x})\bar{\partial_x} u)(x,y) - \partial_y(b(\cdot -\frac{h_2}{2}\vec{e_y})\bar{\partial_y} u)(x,y)$ , $\forall (x,y)\in \Omega_h$,    (12)</center>
<center>$u_h(x,y) = 0$ , $\forall (x,y)\in\partial\Omega_h$</center>
<br>

#### **e) Argumenta, según los resultados vistos en clases de teoría, por qué este método numérico es convergente para el problema presentado (puedes citar sin demostrar cualquier resultado que hayamos dado en teoría, pero debes comprobar que estamos en las hipótesis de los teoremas).**

Para probar la convergencia de este método numérico y como bien se indica en el enunciado, es suficiente asegurar que se cumplen las hipótesis del teorema visto en clase para convergencia de las ecuaciones elípticas.

Estas hipótesis son que $a(x,y), b(x,y)$ son $C^1(\bar{\Omega})$ y que $f(x,y)$ es $C^(\bar{\Omega})$. En primer lugar veamos que se cumple la hipótesis que se requiere para $a(x,y)$. Para $b(x,y)$ sería completamente análogo ya que tienen exactamente la misma forma así que no se hace esa comprobación para no ser repetitivos.

La continuidad en cada uno de los trozos en los que se define $a$ es trivial pues en el primero la función es constantemente 1 y en el segundo tiene forma de exponencial. Se puede comprobar que no hay ningún salto en los "cambios de trozo" (es decir en $x = \pm 1$) ya que los límites laterales en todos los casos valen $1$ (es una comprobación trivial ver que cuando $|x| = 1$ la exponencial vale $1$).

Veamos ahora la diferenciabilidad de la función $a(x,y)$. Por motivos evidentes, solo me centro en estudiar la derivada respecto de $x$. Calculando dicha derivada se obtiene que:

<center>$a'(x) = 0$ : $|x| < 1$</center>
<center>$a'(x) = -2k_x(|x| -1)e^{-k_x(|x|-1)^2}$ : $|x| > 1$</center>
<br>

Claramente ninguno de los dos trozos presenta ningún problema y en los puntos $x = \pm 1$, por ambos lados las derivadas laterales valen 0.

Por todo esto se tiene que la función $a(x,y)$ es continua y derivable, luego se tiene $a\in C^1(\Omega)$ como se quería.

Por último, solo queda comprobar que el término de producción asociado a la estufa es una función continua, lo cual es directo pues como los dos trozos que conforman $f$ con claramente continuos y en el salto en el que $\sqrt{x^2+y^2} = 1/2$ se tiene que la parte con la raíz se anula, se tiene que efectivamente $f \in C^0(\Omega).$

**La expresión de (12) se puede escribir en modo expandido:**

#### <center>$A_hu_h(x,y) = -\frac{a\left(x+\frac{h_1}{2}\right)\left(u\left(x+h_1,y\right) - u(x,y)\right) - a\left(x-\frac{h_1}{2}\right)\left(u(x,y) - u\left(x-h_1,y\right)\right)}{h_1^2} - \frac{b\left(y+\frac{h_2}{2}\right)\left(u\left(x,y + h_2\right) - u(x,y)\right) - b\left(y-\frac{h_2}{2}\right)\left(u(x,y) - u\left(x,y-h_2\right)\right)}{h_2^2}$</center>
<br>

**Además, en el caso en el que nos encontremos en un punto cercano a la frontera, por ejemplo $x = h_1$, podemos incorporar automáticamente la condición de Dirichlet $u(0,y) = 0$, de modo que queda:**

#### <center>$A_hu_h(h_1,y) = -\frac{a\left(\frac{3h_1}{2}\right)\left(u\left(2h_1,y\right) - u(h_1,y)\right) - a\left(\frac{h_1}{2}\right)\left(u(h_1,y) - 0\right)}{h_1^2} - \frac{b\left(y+\frac{h_2}{2}\right)\left(u\left(h_1,y + h_2\right) - u(h_1,y)\right) - b\left(y-\frac{h_2}{2}\right)\left(u(h_1,y) - u\left(h_1,y-h_2\right)\right)}{h_2^2}$</center>

**Si representamos nuestra solución numérica por una matriz $U \in \mathcal M_{M\times N}$ (es decir, únicamente considerando los términos de $\Omega_h$ y no $\partial\Omega_h$) el esquema anterior se puede escribir en forma matricial:**

#### **f) Demuestra que la ecuación $A_hu_h = f$ se puede escribir como:**

#### <center>$T_yU + UT_x = F$    (15)</center>
#### **donde $T_y \in \mathcal M_{M\times M}, T_x \in \mathcal M_{N\times N}$ y $F\in \mathcal M_{M\times N}$**

Para este apartado, se va a usar la notación que ya se ha usado en otras ocasiones. Sabiendo que nuestro mallado tiene $N+2$ y $M+2$ puntos en las variables $x$ e $y$ respectivamente, se va a denotar:

#### <center>$u(x_i,y_j) \equiv u_i^j$</center>
<br>

Donde, evidentemente $i = 0,1,...,N+1$ y $j = 0,1,...,M+1$. Se sigue la misma notación para $a,b,f$.

Siguiendo la expresión expandida de la ecuación (12), vamos a ver cómo podemos escribir dicha expresión siguiendo esta nueva notación y con vistas a llegar a la ecuación matricial. Dicha expresión, reagrupando los términos queda (nos quedamos en la parte interior de nuestro mallado, es decir, quitamos los bordes donde tenemos condiciones Dirichlet nulas, se puede considerar así que para lo que sigue, estamos en $i = 1,2,...,N$ y $j = 1,2,...,M$):

#### <center>$-\frac{1}{h_1^2}\left(a_{i+1/2}u_{i+1}^j - \left(a_{i+1/2} + a_{i-1/2}\right)u_i^j + a_{i-1/2}u_{i-1}^j\right) - \frac{1}{h_2^2}\left(b^{j+1/2}u_{i}^{j+1} - \left(b^{j+1/2} + b^{j-1/2}\right)u_i^j + b^{j-1/2}u_{i}^{j-1}\right) = f_i^j \Longrightarrow$</center>

#### <center>$\Longrightarrow -\frac{1}{h_2^2}\left(0,0,...,0,b^{j-1/2}, -b^{j+1/2} - b^{j-1/2}, b^{j+1/2},0,...,0,0\right)\left(u_i^1, u_i^2, ..., u_i^{j-1}, u_i^j, u_i^{j+1},...,u_i^M\right)^T - \frac{1}{h_1^2} \left(u_1^j, u_2^j, ..., u_{i-1}^{j}, u_i^j, u_{i+1}^{j},...,u_N^j\right)\left(0,0,...,0,a_{i-1/2}, -a_{i+1/2} - a_{i-1/2}, a_{i+1/2},0,...,0,0\right)^T = f_i^j$</center>
<br>

Donde aquí, lo único que se ha hecho es reescribir la expresión vectorialmente (donde $^T$ indica que el vector es columna) llenando de ceros donde sea necesario para que efectivamente la expresión sea totalmente equivalente y así nos vamos acercando a la expresión matricial buscada.

Como comentario, para que quede totalmente claro, el primer elemento no nulo del vector de $b^j$ (es decir, $b^{j-1/2}$) se encuentra en la posición $j-1$ (y queda así multiplicando a $u_i^{j-1}$ y se tiene lo que se quiere) y de forma totalmente análoga, el primer elemento no nulo del vector de $a_i$ ($a_{i-1/2}$) está en la posición $i-1$. Además, para ver cómo las dimensiones van cuadrando con lo que buscamos, el primer sumando tiene vectores de dimensión $M$ y el segundo, de dimensión $N$

Ahora la idea es extender esta expresión vectorial para todos los valores de $i,j$, para ello ya sí usamos una expresión matricial (este paso como digo no es más que generalizar lo anterior para tener una expresión compacta para cada $i,j$). Lo que se obtiene es que, efectivamente, se puede escribir nuestra ecuación como $T_y U + UT_x = F$ con las dimensiones que buscábamos donde, cada una de las matrices es (en el caso de $T_x, T_y$ voy a escribir las matrices sin el término $1/h_1^2$ o $1/h_2^2$, aunque lo dejo claro escribiendo qué es cada matriz:):

<center> $-h_2^2\cdot T_y$ es de la forma: \begin{pmatrix}

<center> $-h_1^2\cdot T_x$ es de la forma: \begin{pmatrix}

<center>$U$ es de la forma: \begin{pmatrix}

Con esto, es evidente que se cumple la expresión matricial buscada y que, además, dadas las dimensiones de los vectores en la expresión anterior, se ha de tener que $T_x \in \mathcal M_{N\times N}, T_y \in \mathcal M_{M\times M}, F\in \mathcal M_{M\times N}, U\in\mathcal M_{M\times N}$

Esto está claro por la forma matricial que hemos dado arriba de $T_x, T_y$ ya que si $(i,j)$ es un punto del mallado, lo que nos interesa es conocer $a_{i-1/2} (\equiv a(i-h_1/2)), a_{i+1/2}$ y $b^{j-1/2}, b^{j+1/2}$. Para hacer esto, lo único que se hace es crear nuestros "linspace's" de $x,y$ y evaluar en los puntos intermedios entre sucesivos del linspace.

```python
def construye_f_a_b(extremos_x,N,extremos_y,M,k_x,k_y,C): #extremos_x, extremos_y simplemente son los puntos entre los que se mueven x,y (-3/2 y 3/2)
    x = np.linspace(extremos_x[0], extremos_x[-1], N+2)
    y = np.linspace(extremos_y[0], extremos_y[-1], M+2)
    
    global h1
    global h2
    h1 = x[1] - x[0]
    h2 = y[1]- y[0]
    
    #Teniendo esto, lo útil es crear una malla (matriz) que es, esencialmente, construir las coordenadas dadas por estos dos vectores
    xx, yy = np.meshgrid(x, y, sparse = True)
    
    #Construcción de f(x,y)
    f = C*(1-2*np.sqrt(xx**2+yy**2)) #Obsérvese que aquí aún no hemos considerado que a partir de norma 1/2 la función vale 0
    f[f<=0] = 0 #La función cambia a ser constantemente cero cuando el primer trozo (tal como se define) llega al cero, es justo lo que pongo aquí
    #Cuando la función de la forma de arriba es menor o igual que cero, sustituimos por todo ceros
    
    #Construcción de a(x,y) y b(x,y)
    #Primeramente pongo las funciones tal y como son para |x| y |y| > 1 y después tengo en cuenta el otro caso:
    #Además, tengo en cuenta que no quiero evaluar en el mallado en sí, si no en los puntos intermedios.
    a = np.exp(-k_x*(np.abs(xx + h1/2)-1)**2 + yy*0) #El añadir este sumando de yy*0 se explica después, pero es por mera utilidad
    b = np.exp(-k_y*(np.abs(yy + h2/2)-1)**2 + xx*0)
    
    #Un comentario aclaratorio es que ahora el último punto de xx+h1/2 e yy+h2/2 queda fuera del mallado, pero ese no se va a necesitar pues...
    #... ya hemos dicho que nos quedamos en la parte interior del mallado, así las dimensiones cuadrarán como deben ya que, por ejemplo para a...
    #... si hay N+2 puntos, se pueden calcular N+1 restas de puntos sucesivos, pero la del final no la queremos, de ahí N restas y dim NxN
    
    #Esta parte puede resultar confusa, prefiero explicarla después del código, pero es para sustituir por 1 donde corresponde:    
    a[:,int(np.argwhere(x<=-1)[-1]):int(np.argwhere(x>=1)[0])] = 1
    b[int(np.argwhere(y<=-1)[-1]):int(np.argwhere(y>=1)[0]),:] = 1
    
    return(f,a,b)
    
    
```

**Breve explicación del código:**

Para hallar las tres funciones, como se ve, lo que se hace es construir el mallado ($xx,yy$ son los arrays de nuestro dominio, pero el hacerlo con meshgrid permite obtener una función bidimensional que es esencialmente una matriz de valores).
La obtención de $f(x,y)$ es bastante sencilla y no creo que merezca mayor explicación, pero la obtención de $a(x,y),b(x,y)$ puede resultar más difícil de comprender si no hago un par de comentarios.

En primer lugar obtengo $a,b$ solo teniendo cuenta el segundo trozo exponencial. Ahora bien, por ejemplo en el caso de $a(x)$ he añadido un sumando que sea "$+yy\cdot 0$" que no cambia nada en cuanto a los valores pero así el programa interpreta bien que es algo bidimensional (si no, al depender solo de xx y ser este un array unidimensional $a$ me saldría como un vector también). Así, tengo una matriz de valores para $a$ y otra para $b$ que solo dependen de $x$ e $y$ respectivamente.

Por último, para tener en cuenta dónde estas funciones valen 1, repestivamente cuando $|x|, |y| \leq 1$, uso la función argwhere de numpy que saca todos los índices de un array que cumplen una cierta condición. Con esto y usando las condiciones de que sea, por ejemplo, $x<-1$ y $x>1$, como los linspace's están ordenadon, nos interesan el último índice de la primera condición (cuando x empieza a ser mayor que -1) y el primero de la segunda (cuando empieza a ser mayor que 1). Todos los valores cuyos índices estén entre estos dos, pasan a valer 1.

**Representación de las tres funciones:**

```python
f,a,b = construye_f_a_b(np.array([-3/2,3/2]), 1000, np.array([-3/2,3/2]), 2000,25,25, 40) #Uso los parámetros que se mostraban arriba

plt.pcolormesh(np.linspace(-3/2,3/2,1002),np.linspace(-3/2,3/2,2002),f, cmap = 'inferno') #Hay 1002 y 2002 puntos en x, y respectivamente
plt.colorbar()
plt.title('f(x,y). Mapa de calor')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.pcolormesh(np.linspace(-3/2,3/2,1002),np.linspace(-3/2,3/2,2002),a, cmap = 'inferno') #Hay 1002 y 2002 puntos en x, y respectivamente
plt.colorbar()
plt.title('a(x,y). Mapa de calor')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.pcolormesh(np.linspace(-3/2,3/2,1002),np.linspace(-3/2,3/2,2002),b, cmap = 'inferno') #Hay 1002 y 2002 puntos en x, y respectivamente
plt.colorbar()
plt.title('b(x,y). Mapa de calor')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Efectivamente lo que se observa concuerda con la explicación dada en el PDF donde está el enunciado completo de este entregable.

En primer lugar, se tiene un término de producción $f$ (estufa) en el centro de la habitación, donde se ve claramente más caliente en nuestro mapa de calor, y que va decayendo a medida que nos alejamos del centro para dar lugar a producción nula a partir de distancia radial al centro 1/2.

Por otro lado, en cuanto a $a,b$ se observa también lo que se quería. Lo más obvio es decir que son funciones simétricas respeto a $y = 0, x = 0$ respectivamente como era de esperar, que $a \equiv a(x)$ y $b \equiv b(y)$.
Además, se puede comentar que se tiene lo que se quería, que son paredes de grosor 1/2 aislantes ya que a partir de $|x|,|y| = 1$, se observa cómo decaen $a$ y $b$ lo que indica una peor conducción del calor en las paredes lo cual concuerda con el hecho de que sean aislantes.

#### **i) Construye las matrices $T_x, T_y, F$ de la ecuación (15). Resuelve el problema elíptico numéricamente resolviendo la ecuación de Sylvester (15) (puedes utilizar un método preprogramado o descomponer en productos de Kronecker) y representa tu solución $U$ en el dominio $\Omega$. ¿Se alcanzan los valores frontera adecuadamente?**

No voy a volver a construir la matriz F otra vez en una nueva función ya que, por la forma que toma la matriz y cómo la construí yo en el apartado anterior, es hacer lo mismo. Lo único que hay que tener en cuenta es que hay que quitar las dos filas y las dos columnas de las fronteras (de cara a que casen las dimensiones).

```python
def construye_Tx_Ty_F(a,h1,b,h2,f): #Estos son los parámetros que necesitamos
    
    #a,b son matrices (solo dependiendo de x e y) de cara a representarlo antes, pero nos interesa ahora trabajar solo con arrays:
    a_arr = a[0,:-1] #Quito último elemento pues como se explicó en la celda de la función anterior, sobra para nuestro cálculo
    b_arr = b[:-1,0] #Igual
    
    #La idea va a ser construir Tx, Ty a partir de np.diag, así que lo primero será crear los arrays de las diagonales:
    #Podría pensarse que es necesario un bucle, pero np.diff nos va a ayudar en la eficiencia computacional:
    primera_diag_a = np.diff(a_arr) - 2*a_arr[1:] #Explicado debajo
    segunda_diag_a = a_arr[1:-1]
    
    primera_diag_b = np.diff(b_arr) - 2*b_arr[1:]
    segunda_diag_b = b_arr[1:-1]
    
    #Creo Tx y Ty
    Tx = -1/h1**2*(np.diag((primera_diag_a)) + np.diag((segunda_diag_a),1) + np.diag((segunda_diag_a),-1))
    Ty = -1/h2**2*(np.diag((primera_diag_b)) + np.diag((segunda_diag_b),1) + np.diag((segunda_diag_b),-1))
    
    #F ya está calculada con la función anterior, quito las fronteras:
    F = f[1:-1,1:-1]
    
    return(Tx,Ty, F)
    
```

Puede ser interesante dar una breve explicación sobre la construcción de $T_x, T_y$, en particular sobre cómo se ha construido su diagonal principal. En primer lugar, por cómo se construyeron las matrices $a,b$ en la función cuya finalidad era esa, al quedarnos solo con una fila (y una columna en el caso de b), lo que se tienen ya son los vectores que nos interesan. Por ejemplo, para $a$ ya se tiene calculado el vector $(a_{1/2}, a_{3/2}, ..., a_{(N+1)+1/2})$. Lo siguiente que se ha hecho es quitar el último elemento de dicho vector que no nos interesa como ya sabemos y, después, con np.diff lo que se obtiene es el siguiente vector (de una dimensión menos lógicamente): $(a_{3/2} - a_{1/2}, ..., a_{N+1/2} - a_{N-1/2})$ a lo que, si se le resta el vector $(2a_{3/2}, 2a_{5/2}, ... , 2a_{N+1/2})$, se obtiene la diagonal principal de $T_x$ (y análogamente de $T_y$)

```python
#Una rápida comprobación de las dimensiones:

Tx, Ty, F = construye_Tx_Ty_F(a,h1,b,h2,f)
print('En el caso que tenemos, N = 1000 y M = 2000. Sabiendo esto: \n')
print('Las dimensiones de la matriz F deberían ser (2000,1000) y son:', np.shape(F), '\n')
print('Las dimensiones de la matriz Tx deberían ser (1000,1000) y son:', np.shape(Tx), '\n')
print('Las dimensiones de la matriz Ty deberían ser (2000,2000) y son:', np.shape(Ty), '\n')
```

Resolución del problema elíptico utilizando la función preprogramada de scipy que se muestra en el enunciado.

```python
U = scipy.linalg.solve_sylvester(Ty,Tx,F)

plt.pcolormesh(np.linspace(-3/2+h1,3/2-h1,1000),np.linspace(-3/2+h2,3/2-h2,2000),U, cmap = 'inferno') 
#Ahora hay 1000 y 2000 puntos al haber quitado fronteras y hay que tener en cuenta que no se puede repreentar en +- 3/2 si no en el siguiente
plt.colorbar()
plt.title('U(x,y). Mapa de calor')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Además, como se pide analizar que los valoes frontera se alcancen correctamente, como se puede observar, la malla de $U(x,y)$ tiende a 0 en los cuatro bordes, lo cual concuerda con el hecho de que en esos bordes haya condiciones de frontera Dirichlet nulas y por tanto la solución se comporta como debería.

De hecho, se puede observar cómo a medida que nos acercamos a $|x|, |y| = 3/2$ el gradiente de la temperatura de la sala aumenta considerablemente precisamente para alcanzar esos valores nulos en las fronteras.
