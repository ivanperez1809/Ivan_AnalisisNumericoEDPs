## **Apartado 3) Modelo simplificado**
(Ver introducción en el guion de la práctica para más detalle)

**Para experimentar un poco, vamos a suponer que la velocidad únicamente depende de la posición $v(x)$. En concreto, supondremos que nuetsra carretera estará en obras en los primeros 50 km, por lo que la velocidad máxima será de 70 km/h. Posteriormente, la velocidad se irá incrementando entre el km 50 y 60 hasta la velocidad máxima de la vía: 120 km/h. El resto del trayecto los coches circularán a dicha velocidad. Escrito en forma matemática:**

**<center>$v_s(x) = 70$ si $x\leq 50$</center>**
**<center>$v_s(x) = 70 + 50\left(1-\frac{60-x}{10}\right)$ si $50\leq x\leq 60$</center>**
**<center>$v_s(x) = 120$ si $x\geq 60$</center>**
<br>

**Para completar el problema debemos dar unas condiciones iniciales y de frontera. La condición inicial vendrá dada por:**

#### **<center>$u_0(x) = e^{-\frac{|x-10|^2}{25}}$</center>**
<br>

**que modeliza una acumulación de coches en torno al km 10. En cuanto a las condiciones frontera, como los coches viajan de izquierda a derecha, bastará con imponerlas en la parte izquierda. Asumiremos que existe una densidad constante de 0.1:**

**<center>$u(0,t) = 0.1$ $\forall t >0 $</center>**
<br>

**Nuestro problema queda, por tanto:**

**<center>$u_t(x,t) + (v_s(x)u(x,t))_x = 0$ : $x\in [0,100]$,  $t\in [0,T]$</center>**
**<center>$u(x,0) = e^{-\frac{|x-10|^2}{25}}$ : $x\in [0,100]$</center>**
**<center>$u(0,t) = 0.1$ : $t\in [0,T]$</center>**

**Para simplificar el problema, podemos considerar el cambio de variable $\theta(x,t) = v_s(x)u(x,t)$**

#### **a) Demuestra que con el cambio de variable el problema queda:**

**<center>$\theta_t(x,t) + v_s(x)\theta_x(x,t) = 0$ : $x\in [0,100]$,  $t\in [0,T]$</center>**
**<center>$\theta(x,0) = v_s(x)e^{-\frac{|x-10|^2}{25}}$ : $x\in [0,100]$</center>**
**<center>$\theta(0,t) = 7$ : $t\in [0,T]$</center>**
<br>

Para ver que haciendo este cambio de variable se obtiene dicho problema de contorno, simplemente hay que comprobar que se satisface la EDP y las condiciones de contorno.

Empecemos viendo que se cumple la EDP, para ello calculamos las derivadas parciales en cuestión:

La derivada temporal queda: 
<center>$\theta_t(x,t) = v_s(x)u_t(x,t)$</center>
<br>
mientras que, para la parte espacial, si no se desarrolla la regla del producto se obtiene que: 
<center>$\theta_x(x,t) = (v_s(x)u(x,t))_x$</center>
<br>

Con esto, si se calcula 
<center>$\theta_t(x,t) + v_s(x)\theta_x(x,t) = v_s(x)u_t(x,t) + v_s(x)(v_s(x)u(x,t))_x = v_s(x)\cdot (u_t(x,t) + (v_s(x)u(x,t))_x) = 0$</center>
<br>
Donde, en la última igualdad se ha usado la ecuación que sabemos que se cumple para $u(x,t)$. Por tanto, ya se tiene que efectivamente
<center>$\theta_t(x,t) + v_s(x)\theta_x(x,t) = 0$ si $x\in [0,100]$,  $t\in [0,T]$</center>
<br>

La condición inicial se cumple de forma trivial pues:
<center>$\theta(x,0) = v_s(x)u(x,0) = v_s(x)e^{-\frac{|x-10|^2}{25}}$ si $x\in [0,100]$</center>
<br>

Y la condición de frontera en la izquierda es igual de sencilla:
<center>$\theta(0,t) = v_s(0)u(0,t) = 70 \cdot 0.1 = 7$ si $t\in [0,T]$</center>


**El problema así planteado es sencillo de resolver con la teoría que hemos visto en clase. Como $v_s$ no depende de $\theta$, el problema es lineal y las curvas características son independientes del dato inicial.**

#### **b) Pinta las curvas características (del problema para $\theta$) en el plano $x-t$ para $x\in[0,100]$ y $t\in [0,1]$ (hazlo con el ordenador). ¿Qué observas?**

Para hacer este apartado, en primer lugar veamos cómo son dichas curvas características. Denotándolas por $x(s),t(s)$ como es costumbre se obtiene de forma sencilla que:

<center>$x'(s) = v_s(x)$ , $t'(s) = 1$</center>
<br>

De donde se deduce fácilmente que $t = s$ (pues suponemos $t(0) = 0$)y para la parte en $x$ hay que tener en cuenta la forma que hemos visto que tiene $v_s(x)$.

Para este cálculo, aunque no lo haya comentado explícitamente, se ha utilizado que la solución $\theta$ es constante a lo largo de las curvas características, es decir $d\theta/ds = 0$ y esto, tal y como ya hemos visto varias veces, usando la regla de la cadena permite igualar los términos tal y como lo he hecho.

<center>Si $x\leq 50$ se tiene: $x'(s) = 70$</center>
<center>Si $50 \leq x\leq 60$ se tiene: $x'(s) = 70 + 50\left(1-\frac{60-x(s)}{10}\right) = 5x(s) - 180$</center>
<center>Si $x\geq 60$ se tiene: $x'(s) = 120$</center>
<br>

Si denotamos por $x_0^{(1)}, x_0^{(2)}, x_0^{(3)}$ a los puntos iniciales en cada uno de los trozos y dado que ya se ha visto que $t = s$, se tiene que:

<center>$x(t) = x_0^{(1)} + 70t$ si $x\leq 50$</center>
<center>$x(t) = \left(x_0^{(2)} - 36\right)e^{5t} + 36$ si $50\leq x \leq 60$</center>
<center>$x(t) = x_0^{(3)} + 120t$ si $60\leq x$</center>
<br>

Si se mira la siguiente celda, verá que se ha hecho ahí la segunda EDO con sympy (las otras dos son triviales y no se da más detalle) y la constante se obtiene sin más que tomar $s = 0$.

Además, comentar el hecho de diferenciar entre los datos iniciales $x_0^{(i)}$ ya que dependiendo del punto en $x$ inicial, la EDO para la curva característica es una u otra y como en los tres casos son crecientes en $x$ se llega un momento en el que se ha de cambiar de "trozo".
Por ilustrarlo mejor, imaginemos que se considera una curva empezando en $x=0$, entonces la curva va a evolucionar siguiendo la recta del primer trozo hasta llegar a $x = 50$, donde seguirá evolucionando siguiendo la exponencial del segundo (con $x_0^{(2)} = 50$ lógicamente, ya que es el punto inicial en $x$ para ese trozo). Desde ahí seguirá creciendo hasta llegar a $x = 60$ y finalmente cambia al último trozo. Si por ejemplo el punto inicial fuera $x = 55$ entonces ya directamente empezará en el segundo trozo, etc.

Por supuesto, aquí lo importante también va a ser obtener los tiempos $t_f^{(i)}$ (siguiendo la misma notación) en los que se cambia de trozo. Lo explico con todo detalle aquí para que en el código se siga sin problemas:

**Si $x_0 < 50$ entonces:**
<center>$50 = x_0 + 70t_f^{(1)} \Longrightarrow t_f^{(1)} = \frac{50-x_0}{70}$</center>
<br>

En este instante temporal se cambia al segundo trozo y permanece ahí un intervalo de tiempo $\Delta t_2$ hasta que se cumple:

<center>$60 = (50 - 36)e^{5\Delta t_2} + 36 \Longrightarrow \frac{12}{7} = e^{5\Delta t_2} \Longrightarrow \Delta t_2 = \frac{1}{5}ln\left(\frac{12}{7}\right)$</center>
<br>

Como se verá en el código, lo importante para cada trozo es cuánto tiempo este ahí para calcular bien $x(t)$, pero para representarlo importante entre qué tiempos se dé ese intervalo.

Por tanto, 
<center>$t_f^{(2)} = t_f^{(1)} + \Delta t_2 = \frac{50 - x_0}{70} + \frac{1}{5}ln\left(\frac{12}{7}\right)$</center>
<br>

Razonando de forma totalmente análoga, en ese tiempo se cambiaría en el tercer trozo en el que permanecería un intervalo de tiempo $\Delta t_3 = 1/3$ (el cálculo es muy sencillo) y de ahí:

<center>$t_f^{(3)} = \frac{50 - x_0}{70} + \frac{1}{5}ln\left(\frac{12}{7}\right) + \frac{1}{3}$</center>
<br>

**Si $x_0 \in [50,60]$:**

El razonamiento es completamente análogo, se empieza en el segundo trozo y se tiene:

<center>$60 = (x_0 - 36)e^{5t_f^{(2)}} + 36 \Longrightarrow t_f^{(2)} = \frac{1}{5}ln\left(\frac{24}{x_0-36}\right)$</center>
<br>

Ahí cambia de trozo y en el último permanece el mismo tiempo que en el caso anterior con lo que:

<center>$t_f^{(3)} = \frac{1}{5}ln\left(\frac{24}{x_0-36}\right) + \frac{1}{3}$</center>
<br>

**Si $x_0 \geq 60$:**

<center>$t_f^{(3)} = \frac{100-x_0}{120}$</center>
<br>

```python
s = sympy.symbols('s')
x = sympy.Function('x')
sympy.dsolve(sympy.Derivative(x(s),s) - 5*x(s) + 180, x(s))
```

Con todo esto, ya se define una función que pinte nuestras curvas características:

```python
def CurvasCaracteristicas(x0): #Solo necesitamos saber en qué punto queremos empezar la curva tal y como se ha explicado
    
    #Lo más cómodo que se me ocurre es poner condicionales dependiendo del valor de x0 para no complicar demasiado el código
    if x0<50:
        t1 = np.linspace(0,(50-x0)/70,100) #Primer intervalo temporal
        x1 = x0 + 70*t1
        
        t2 = np.linspace(t1[-1], t1[-1] + 1/5*np.log(12/7), 100) #Segundo intervalo temporal
        Delta_t2 = np.linspace(0,1/5*np.log(12/7), 100)
        x2 = 14*np.exp(5*Delta_t2) + 36 #Para que los extremos cuadren, solo tengo que tener en cuenta el Delta_t2 (explicado arriba)
        
        t3 = np.linspace(t2[-1], t2[-1] + 1/3, 100)
        Delta_t3 = np.linspace(0,1/3,100)
        x3 = 60 + 120*Delta_t3
        
        return(np.concatenate((x1,x2,x3)), np.concatenate((t1,t2,t3)))
        
    elif 50<=x0<=60:
        t2 = np.linspace(0, 1/5*np.log(24/(x0-36)), 100)
        x2 = (x0 - 36)*np.exp(5*t2) + 36
        
        t3 = np.linspace(t2[-1], t2[-1] + 1/3, 100)
        Delta_t3 = np.linspace(0, 1/3, 100)
        x3 = 60 + 120*Delta_t3
        
        return(np.concatenate((x2,x3)), np.concatenate((t2,t3)))
    
    else:
        t3 = np.linspace(0,(100-x0)/120, 100)
        x3 = x0 + 120*t3
        return(x3, t3)
```

```python
for x0 in np.arange(0,100,2):
    x,t = CurvasCaracteristicas(x0)
    plt.plot(x,t, color = 'black')

plt.title('Curvas características modelo simplificado')
plt.xlabel('x')
plt.ylabel('t')
plt.xlim([0,100])
plt.ylim([0,1])
plt.show()
```

Lo primero que se observa y que es importante comentar es que las curvas características no se cortan y las pendientes se comportan como deberían según las expresiones que hemos dado para $(x(s),t(s))$ para los distintos trozos.

#### **c) Utiliza un esquema upwind de primer orden para resolver el sistema en $\theta$. ¿Cuál es, en este caso la condición CFL?**

Siguiendo los razonamientos que se han empleado en otras ocasiones cuando se realizan este tipo de m´´etodos explícitos, se va a considerar un mallado espaciotemporal de salto espacial $\Delta x$ y temporal $\Delta t$ dejando un mallado de puntos $(x_i, t_j)$. Así, se denotará a la solución en dichos puntos por $\theta(x_i,t_j) \equiv \theta_i^j$.

Ahora bien, según lo visto en clase (ver PDF _Esquemas Numéricos Hiperbólicos_), el esquema upwind queda de la siguiente manera:

#### <center>$\frac{\theta_i^{j+1} - \theta_i^j}{\Delta t} + a\frac{\theta_i^j - \theta_{i-1}^j}{\Delta x} = 0 \Longrightarrow $</center>

#### <center>$\Longrightarrow \theta_i^{j+1} = \theta_i^j - a\frac{\Delta t}{\Delta x}\left(\theta_i^j - \theta_{i-1}^j\right)$</center>
<br>

Ahora bien, he seguido la notación de lo que se ha visto en clase, pero en este caso hay que tener en cuenta que $a = v_s(x)$, con lo que realmente se tiene:

#### <center>$\theta_i^{j+1} = \theta_i^j - v_s(x_i)\frac{\Delta t}{\Delta x}\left(\theta_i^j - \theta_{i-1}^j\right)$</center>
<br>

Reordenando de cara a usar esta fórmula en el código se tiene que:

#### <center>$\theta_i^{j+1} = \theta_i^j\left(1 - v_s(x_i)\frac{\Delta t}{\Delta x}\right) + v_s(x_i)\frac{\Delta t}{\Delta x}\theta_{i-1}^j$</center>
<br>

En este caso, comentando ahora la condición CFL que nos aparece, es que $v_s(x)\cdot \frac{\Delta t}{\Delta x} \leq 1$. Ahora bien, obsérvese que la función $v_s(x)$ es creciente en $x$ y de valor máximo $120$ con lo que se puede escribir esta condición como:

<center>$120\cdot \frac{\Delta t}{\Delta x} \leq 1$</center>
<br>

Por cómo voy a construir el mallado (voy a usar linspaces con lo cual voy a controlar el número de puntos del mallado y no tanto los saltos espaciotemporales), conviene dar la siguiente indicación y es que si se toman $N+1$ puntos en la parte espacial y $M+1$ en la parte temporal se tiene que $\Delta x = 100/N$ y $\Delta t = 1/M$ con lo cual la condición CFL se puede reescribir en términos de $N$ y $M$ como:

<center>$\frac{\Delta t}{\Delta x} \leq 120 \Longrightarrow \frac{N}{100M} \leq 120 \Longrightarrow M \geq 1.2N$</center>

```python
#Definimos una función que nos devuelva la solución como ya se ha hecho varias veces en el curso con este tipo de cosas:
def matrizSolucion(x,t): 
    #La idea va a ser crear la matriz para cualquier mallado x,t y a posteriori llamar a esta con uno que cumpla la CFL
    #x,t suponemos que son linspaces (Delta_t, Delta_x constantes) que cumplan dicha condición CFL
    
    #Lo primero, construyo v_s(x) que es necesaria para este esquema:
    v = np.zeros(len(x)) #Es a trozos
    fin_trozo1 = len(x[x<50]) ; fin_trozo2 = len(x[x<=60])
    v[:fin_trozo1] = 70 ; v[fin_trozo1:fin_trozo2] = 70 + 50*(1-(60-x[fin_trozo1:fin_trozo2])/10) ; v[fin_trozo2:] = 120
    
    #Lo segundo, hallo Delta_t y Delta_x
    Delta_t = t[1] - t[0] #Espaciado constante
    Delta_x = x[1] - x[0]
    lambda_ = Delta_t/Delta_x #Siguiendo la notación de clase lambda = delta_t/delta_x
    
    #Construye las CI y las CC ya que dependen de v
    CI = v*np.exp(-np.abs(x-10)**2/25)
    CC = 7*np.ones(len(t))
    
    matriz = np.zeros((len(t), len(x))) #Inicializamos ahí
    matriz[0,:] = CI #Aquí ya se está suponiendo que CI es de la misma longitud que el vector de tiempos
    matriz[:,0] = CC
    
    #Bucle para "rellenar la matriz"
    for i in range(1,len(t)-1): 
        matriz[i,1:] = matriz[i-1,1:]*(1-lambda_*v[1:]) + lambda_*v[1:]*matriz[i-1,:-1]
    
    return (matriz,v)
```

```python
#Tomamos x,t tales que cumplan la condición CFL
x = np.linspace(0,100,1001) # N = 1000
t = np.linspace(0,1,1201) # M = 1200 y se cumple la condición CFL (de hecho se da la igualdad)

matriz_theta = matrizSolucion(x,t)[0]

plt.pcolormesh(x,t,matriz_theta, cmap = 'inferno')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('t')
plt.title(r'$\theta$ (x,t) para el modelo simplificado')
plt.show()
```

Aunque ya se ha comentado lo que se necesitaba de la condición CFL simplemente destacar que he usado el límite en el que se da la igualdad mostrada arriba. Esto no tendría por qué hacerse así, pero conseguimos que al coger $N = 1000$ sea un mallado lo suficientemente fino pero no hacerlo demasiado fino en la parte temporal salvando así cierta eficiencia computacional.

## **Apartado 4) Modelo no lineal**

**Ahora vamos a hacer nuestro modelo un poco más realista. Para ello, esta vez vamos a considerar que la velocidad sí depende de la densidad de coches. Consideraremos que la carretera no está en obras y esta vez se podrá circular a 120 km/h todo el trayecto. Sin embargo, los conductores tratarán de ser precavidos y si ven que hay mucha densidad de coches delante de ellos, reducirán su velocidad. En concreto, esta vendrá dada por:**

**<center>$v_{nl}(u) = 120(1-u)$</center>**
<br>

**Así, la ecuación de conservación quedará:**

**<center>$u_t + 120(u(1-u))_x = 0$</center>**
<br>

**En este caso el problema es más complejo, ya que no es lineal y las curvas características dependen del dato inicial. Como dato inicial se va a considerar:**

**<center>$u_0(x) = 0.2 + 0.5e^{-\left(\frac{|x-50|}{30}\right)^2}$</center>**
<br>

**Que representará de nuevo una acumulación de coches en torno al km 50**

#### **g) Pinta también estas curvas características y comprueba que se cortan**

En primer lugar, por comodidad, desarrollo un poco la ecuación de conservación que tenemos ahora:
<center>$u_t + 120u_x(1-u) -120uu_x = u_t + 120u_x(1-2u) = 0$</center>
<br>

El razonamiento para obtener las curvas características es completamente análogo a lo hecho con el modelo simplificado. De nuevo las denotamos por $(x(s),t(s))$. Sin más que identificar términos como hacíamos antes (usando que la solución $u$ es constante en nuestras curvas características) en nuestra nueva ecuación de conservación (no ahondo más en ello pues se hace siempre igual) se obtiene que:

<center>$t'(s) = 1$ ; $x'(s) = 120(1-2u)$</center>
<br>

De nuevo se tiene de forma trivial que $t = s$ (pues suponemos que empezamos en tiempo 0) y para la otra parte usamos fuertemente el hecho de que $u(x(s),t(s)) \equiv u(x_0,0))$, con lo que $x'(s)$ es constante ya que la $u$ lo es en estos casos. Resumiendo:

<center>$t = s$ ; $x(s) = x_0 + 120(1-2u_0(x_0))t$</center>
<br>

En este caso, como el dato inicial es conocido, la representación de estas curvas no es tan enrevesada como antes en las que teníamos que ir a trozos:

```python
def CurvasCaracteristicas2(x0):
    
    t = np.linspace(0,100) #Solo un vector de tiempos, nada extraño
    x = x0 + 120*(1-2*(0.2+0.5*np.exp(-(np.abs(x0-50)/30)**2)))*t #Sencillamente usando la expresión mostrada
    
    return(x,t)
```

```python
for x0 in np.arange(0,100,2):
    x_curvas2, t_curvas2 = CurvasCaracteristicas2(x0)
    plt.plot(x_curvas2, t_curvas2, color = 'black')

plt.title('Curvas características modelo no lineal')
plt.xlabel('x')
plt.xlim([0,100])
plt.ylim([0,1])
plt.ylabel('t')
plt.show()
```

#### **h) Plantea el método de Lax Friedrich para este problema y resuelve el problema para un grid con $\Delta x = 0.2$ y $\Delta t = \frac{\Delta x}{120}$. Realiza un vídeo de la solución para $t$ entre 0 horas y 2'5 horas (unas 1500 iteraciones)**

Tal y como se muestra en el guion de la práctica, si uno hace un mallado para nuestra solución $u(x_i,t_n) \equiv u_i^n$ con un paso espacial $\Delta x$ y temporal $\Delta t$ se tiene que (uso que $f_i^n = 120u_i^n(1-u_i^n)$):

<center>$u_i^{n+1} = \frac{1}{2}\left(u_{i+1}^n + u_{i-1}^n\right) - \frac{\Delta t}{2\Delta x}\left(120u_{i+1}^n(1-u_{i+1}^n) - 120u_{i-1}^n(1-u_{i-1}^n)\right) \Longrightarrow$</center>
<br>

<center>$\Longrightarrow u_i^{n+1} = \frac{1}{2}\left(u_{i+1}^n + u_{i-1}^n\right) - \frac{60\Delta t}{\Delta x}\left(u_{i+1}^n(1-u_{i+1}^n) - u_{i-1}^n(1-u_{i-1}^n)\right)$</center>
<br>

Tenemos por tanto, como ya se ha hecho otras veces en la asignatura, una forma de rellenar nuestra matriz solución a partir de las condiciones iniciales y de contorno. Vamos con el programa:

```python
def solLF(delta_x, delta_t): #A diferencia de lo de antes, como ahora me dan los pasos espacio temporales fijos, trabajo con ellos:
    
    #Creo los vectores de x y t
    global x_LF,t_LF
    x_LF = np.arange(0, 100 + delta_x, delta_x)
    t_LF = np.arange(0, 2.5 + delta_t, delta_t)
    
    #Creo los vectores de condiciones iniciales y de contorno
    CC = 0.2*np.ones(len(t_LF))
    CI = 0.2 + 0.5*np.exp(-(np.abs(x_LF-50)/30)**2)
    
    u = np.zeros((len(t_LF),len(x_LF))) #Creo la matriz de ceros para empezar a rellenarla
    
    u[0,:] = CI #Igual con CI
    u[:,0] = CC
    u[:,-1] = CC #Rellenamos las CC
    
    
    for i in range(1,len(t_LF)): #Hago doble bucle, quizá se podría optimizar, pero el tiempo de entrega apremia!!
        for j in range(1, len(x_LF)-1):
            u[i,j] = 1/2*(u[i-1,j+1] + u[i-1,j-1]) - 60*delta_t/delta_x * (u[i-1,j+1]*(1-u[i-1,j+1]) - u[i-1,j-1]*(1-u[i-1,j-1]))
    
    return u
```

```python
delta_x = 0.2
delta_t = delta_x/120

matrizSolucion = solLF(delta_x, delta_t)

plt.pcolormesh(x_LF,t_LF,matrizSolucion, cmap = 'inferno')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('t')
plt.title(r'$u$ (x,t) para el modelo no lineal. Lax-Friedrich')
plt.show()
```
